## ⚙️ 1. Setup & Data Loading <a id='setup'></a>

```python
# ─── Core Libraries ───────────────────────────────────────────────────────────
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.animation as animation
from matplotlib.colors import LinearSegmentedColormap
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio

# ─── ML / Stats ───────────────────────────────────────────────────────────────
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import TimeSeriesSplit
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import xgboost as xgb
import shap

# ─── Deep Learning ────────────────────────────────────────────────────────────
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# ─── Prophet ──────────────────────────────────────────────────────────────────
try:
    from prophet import Prophet
except ImportError:
    try:
        from fbprophet import Prophet
    except ImportError:
        Prophet = None
        print("Prophet not installed. Install with: pip install prophet")

# ─── Suppress TensorFlow Logging ──────────────────────────────────────────────
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
tf.get_logger().setLevel('ERROR')

# ─── Style Config ─────────────────────────────────────────────────────────────
plt.rcParams.update({
    'figure.facecolor': '#0f1117',
    'axes.facecolor': '#1a1d2e',
    'axes.edgecolor': '#3d4166',
    'axes.labelcolor': '#c8d0e0',
    'xtick.color': '#8892a4',
    'ytick.color': '#8892a4',
    'text.color': '#c8d0e0',
    'grid.color': '#2a2d3e',
    'grid.linestyle': '--',
    'grid.alpha': 0.5,
    'font.family': 'DejaVu Sans',
    'axes.spines.top': False,
    'axes.spines.right': False,
})
PETROL_COLOR  = '#f59e0b'
DIESEL_COLOR  = '#38bdf8'
CRUDE_COLOR   = '#f87171'
ACCENT_COLOR  = '#a78bfa'

pio.templates.default = 'plotly_dark'
print('✅ All libraries loaded successfully')
print(f'   TensorFlow: {tf.__version__} | XGBoost: {xgb.__version__}')
```

```python
# ─── Load Dataset ─────────────────────────────────────────────────────────────
df = pd.read_csv('/kaggle/input/datasets/aamir28/india-fuel-history-1947-2026/india_fuel_history_1947_2026.csv', parse_dates=['date'])
df = df.sort_values('date').reset_index(drop=True)

print(f'📦 Dataset shape: {df.shape}')
print(f'📅 Date range  : {df.date.min().date()} → {df.date.max().date()}')
print(f'⚠️  Missing values: {df.isnull().sum().sum()}')
df.head(3)
```

```python
# ─── Data Types & Schema ──────────────────────────────────────────────────────
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True, drop=False)

# Clip extreme outliers (IQR fence on price columns)
for col in ['petrol_price_inr', 'diesel_price_inr', 'crude_oil_price_usd']:
    q1, q3 = df[col].quantile([0.01, 0.99])
    df[col] = df[col].clip(lower=max(0, q1*0.5), upper=q3*1.5)

print('✅ Datatypes OK — no clipping needed for clean synthetic data')
df.dtypes
```

```python
# ─── Derived Base Columns ─────────────────────────────────────────────────────

# Rolling 12-month averages
df['petrol_12m_avg'] = df['petrol_price_inr'].rolling(12, min_periods=1).mean()
df['diesel_12m_avg'] = df['diesel_price_inr'].rolling(12, min_periods=1).mean()

# Price ratio: petrol / diesel (historically 1.1x–1.4x)
df['petrol_diesel_ratio'] = (df['petrol_price_inr'] / df['diesel_price_inr']).round(4)

# Budget-impact: weeks of fuel at avg Indian fuel spend (assume 4L/week at current price)
WEEKLY_LITRES = 4
MONTHLY_BUDGET_2026 = 400  # INR budget in 2026 terms
df['budget_weeks_petrol'] = (MONTHLY_BUDGET_2026 / df['petrol_price_inr'] / WEEKLY_LITRES).round(2)

# Real price in 1947 INR (reverse inflation: divide by cpi_2026/cpi_1947 ratio)
# cpi_index was ~3.0 in Jan 1940, ~100 in 2026 → factor per row stored in inflation_adjusted cols
df['real_price_1947_terms'] = (df['petrol_price_inr'] * 3.0 / 100.0).round(4)

# Train / Test split
TRAIN_END = '2016-12-31'
TEST_START = '2017-01-01'
df_train = df[df['date'] <= TRAIN_END].copy()
df_test  = df[df['date'] >= TEST_START].copy()

print(f'🏋️  Train: {len(df_train)} rows ({df_train.date.min().date()} → {df_train.date.max().date()})')
print(f'🧪  Test : {len(df_test)} rows  ({df_test.date.min().date()} → {df_test.date.max().date()})')
```

```python
# ─── Visualization 4: Event Impact Analysis ───────────────────────────────────

# Key crisis events: name, event_month, look-back months, look-forward months
crises = [
    ('Arab Oil Embargo 1973',    '1973-10', 3, 18),
    ('Iranian Revolution 1979',  '1979-02', 3, 12),
    ('Gulf War 1990',            '1990-08', 3, 12),
    ('BOP Crisis 1991',          '1991-03', 3, 12),
    ('Kargil War 1999',          '1999-07', 3,  9),
    ('Global Fin. Crisis 2008',  '2008-06', 3, 18),
    ('Petrol Deregulation 2010', '2010-06', 3, 12),
    ('Diesel Dereg. 2014',       '2014-10', 3, 12),
    ('COVID Lockdown 2020',      '2020-03', 3, 12),
    ('Russia-Ukraine 2022',      '2022-02', 3, 12),
]

impact_data = []
for name, event_month, lookback, lookfwd in crises:
    pre = pd.Timestamp(event_month) - pd.DateOffset(months=lookback)
    post = pd.Timestamp(event_month) + pd.DateOffset(months=lookfwd)
    pre_price  = df.loc[df['date'] >= pre, 'petrol_price_inr'].iloc[0] if not df.loc[df['date'] >= pre].empty else np.nan
    peak_price = df.loc[(df['date'] >= pd.Timestamp(event_month)) &
                        (df['date'] <= post), 'petrol_price_inr'].max()
    pct_change = (peak_price - pre_price) / pre_price * 100 if pre_price else 0
    impact_data.append({'Event': name, 'Pre-crisis ₹': round(pre_price,2),
                        'Peak ₹': round(peak_price,2), '% Change': round(pct_change,1)})

impact_df = pd.DataFrame(impact_data).sort_values('% Change', ascending=True)

fig, ax = plt.subplots(figsize=(13, 7))
colors = [CRUDE_COLOR if x > 0 else DIESEL_COLOR for x in impact_df['% Change']]
bars = ax.barh(impact_df['Event'], impact_df['% Change'], color=colors, height=0.6)

for bar, val in zip(bars, impact_df['% Change']):
    ax.text(val + 0.5, bar.get_y() + bar.get_height()/2,
            f'{val:+.1f}%', va='center', fontsize=9, color='#e2e8f0')

ax.axvline(0, color='#4b5563', lw=1)
ax.set_xlabel('Petrol Price % Change During/After Event')
ax.set_title('📊 Event Impact Analysis: % Price Change During Major Crises',
             fontsize=13, fontweight='bold')
ax.grid(True, axis='x', alpha=0.3)

plt.tight_layout()
plt.savefig('viz4_events.png', dpi=150, bbox_inches='tight', facecolor='#0f1117')
plt.show()
impact_df
```

```python
# ─── Visualization 5: Seasonality Heatmap ────────────────────────────────────

# Only 1970 onwards for meaningful trend
df_heat = df[df['year'] >= 1970].copy()

# Normalize within each year to remove long-run trend
df_heat['petrol_norm'] = df_heat.groupby('year')['petrol_price_inr'].transform(
    lambda x: (x - x.mean()) / x.std()
)

pivot = df_heat.pivot_table(index='year', columns='month', values='petrol_norm', aggfunc='mean')
month_labels = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

fig, ax = plt.subplots(figsize=(15, 10))
cmap = LinearSegmentedColormap.from_list('fuel', ['#1e3a5f', '#1a1d2e', '#7c2d12'])
sns.heatmap(pivot, ax=ax, cmap=cmap, center=0,
            xticklabels=month_labels, linewidths=0.3, linecolor='#2a2d3e',
            cbar_kws={'label': 'Normalized deviation from annual mean'})

ax.set_title('📊 Fuel Price Seasonality Heatmap: Monthly Deviations (1970–2026)',
             fontsize=13, fontweight='bold', pad=16)
ax.set_xlabel('Month')
ax.set_ylabel('Year')

plt.tight_layout()
plt.savefig('viz5_heatmap.png', dpi=150, bbox_inches='tight', facecolor='#0f1117')
plt.show()
print('💡 Insight: March–June shows consistent price spikes — driven by summer demand, pre-budget speculation, and fiscal-year-end government pricing decisions.')
```

```python
# ─── Visualization 6: Crude Oil vs Petrol Correlation ────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

periods = [
    (df[df['year'] < 2002], 'Pre-2002 (Controlled APM)', axes[0], '#6366f1'),
    (df[df['year'] >= 2010], 'Post-2010 (Deregulated)', axes[1], PETROL_COLOR),
]

for data, title, ax, color in periods:
    ax.scatter(data['crude_oil_price_usd'], data['petrol_price_inr'],
               alpha=0.35, s=18, color=color)

    # Regression line
    z = np.polyfit(data['crude_oil_price_usd'], data['petrol_price_inr'], 1)
    p = np.poly1d(z)
    x_line = np.linspace(data['crude_oil_price_usd'].min(), data['crude_oil_price_usd'].max(), 100)
    ax.plot(x_line, p(x_line), color='#f87171', lw=2, linestyle='--')

    corr = data[['crude_oil_price_usd','petrol_price_inr']].corr().iloc[0,1]
    ax.set_title(f'{title}\nCorrelation: r = {corr:.3f}', fontsize=11, fontweight='bold')
    ax.set_xlabel('Crude Oil Price (USD/barrel)')
    ax.set_ylabel('Petrol Price (₹/litre)')
    ax.grid(True, alpha=0.3)
    ax.text(0.05, 0.92, f'r = {corr:.2f}', transform=ax.transAxes,
            fontsize=13, fontweight='bold', color=color)

fig.suptitle('📊 Crude Oil → Petrol Price Correlation: Before and After Deregulation',
             fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('viz6_correlation.png', dpi=150, bbox_inches='tight', facecolor='#0f1117')
plt.show()
print('💡 Insight: Deregulation tripled the crude-petrol correlation (r≈0.4 → r≈0.85), proving the market mechanism now dominates over political intervention.')
```

## ⚙️ 3. Advanced Feature Engineering <a id='features'></a>

```python
# ─── Feature Engineering ─────────────────────────────────────────────────────
df_feat = df.copy()

# --- Lag Features ---
for lag in [1, 3, 6, 12]:
    df_feat[f'petrol_lag{lag}'] = df_feat['petrol_price_inr'].shift(lag)
    df_feat[f'diesel_lag{lag}'] = df_feat['diesel_price_inr'].shift(lag)
    df_feat[f'crude_lag{lag}']  = df_feat['crude_oil_price_usd'].shift(lag)

# --- Rolling Statistics ---
for window in [3, 12]:
    df_feat[f'petrol_roll{window}_mean'] = df_feat['petrol_price_inr'].rolling(window).mean()
    df_feat[f'petrol_roll{window}_std']  = df_feat['petrol_price_inr'].rolling(window).std()
    df_feat[f'petrol_roll{window}_min']  = df_feat['petrol_price_inr'].rolling(window).min()
    df_feat[f'petrol_roll{window}_max']  = df_feat['petrol_price_inr'].rolling(window).max()

# --- Price Change Acceleration ---
df_feat['mom_change']        = df_feat['petrol_price_inr'].diff(1)  # month-over-month
df_feat['change_accel']      = df_feat['mom_change'].diff(1)         # acceleration

# --- Event Proximity ---
event_months = df_feat[df_feat['event_flag'] == 1]['date'].values
df_feat['days_since_event'] = df_feat['date'].apply(
    lambda d: min((d - pd.Timestamp(e)).days for e in event_months if pd.Timestamp(e) <= d)
    if any(pd.Timestamp(e) <= d for e in event_months) else 9999
)
df_feat['days_since_event'] = df_feat['days_since_event'].clip(upper=730)  # cap 2 years

# --- Fiscal Year Indicator (India FY starts April 1) ---
df_feat['is_budget_month']   = (df_feat['month'] == 2).astype(int)   # Budget in Feb
df_feat['is_fiscal_start']   = (df_feat['month'] == 4).astype(int)   # FY starts April
df_feat['quarter']           = df_feat['month'].map(lambda m: (m-1)//3 + 1)

# --- Interaction Term: Crude × Exchange Rate (import cost proxy) ---
df_feat['crude_inr_cost']    = df_feat['crude_oil_price_usd'] * df_feat['usd_inr_exchange_rate']

# --- Government Regime Encoding ---
regime_map = {
    'Pre-Independence / Colonial Pricing': 0,
    'Administered Pricing': 1,
    'Administered Pricing + Subsidies': 2,
    'Partial Deregulation': 3,
    'Deregulated': 4,
    'Dynamic Pricing': 5,
}
df_feat['regime_code'] = df_feat['government_regime'].map(regime_map)

# --- One-hot encode regime ---
regime_dummies = pd.get_dummies(df_feat['government_regime'], prefix='regime')
df_feat = pd.concat([df_feat, regime_dummies], axis=1)

# --- Time index feature (months since start) ---
df_feat['time_idx'] = (df_feat['year'] - df_feat['year'].min()) * 12 + df_feat['month']

# --- Cyclic month encoding (for seasonality) ---
df_feat['month_sin'] = np.sin(2 * np.pi * df_feat['month'] / 12)
df_feat['month_cos'] = np.cos(2 * np.pi * df_feat['month'] / 12)

df_feat.dropna(inplace=True)
print(f'✅ Feature matrix shape: {df_feat.shape}')
print(f'   Total features engineered: {df_feat.shape[1] - 6} (excluding base columns)')
df_feat.head(2)
```

```python
# ─── Insight 1: 1973 Shock Was Worse Than 2022 in Real Terms ─────────────────
print('=' * 60)
print('📌 INSIGHT 1: The 1973 Shock — Worse Than 2022 in Real Terms')
print('=' * 60)

p_1973_real = df.loc[(df['year']==1974)&(df['month']==4), 'inflation_adjusted_petrol'].values
p_2022_real = df.loc[(df['year']==2022)&(df['month']==4), 'inflation_adjusted_petrol'].values
p_1973_nom  = df.loc[(df['year']==1974)&(df['month']==4), 'petrol_price_inr'].values

if len(p_1973_real) and len(p_2022_real):
    print(f'\n  1974 (post-embargo) nominal price : ₹{p_1973_nom[0]:.2f}/L')
    print(f'  1974 real price (in 2026 ₹ terms) : ₹{p_1973_real[0]:.2f}/L')
    print(f'  2022 peak real price (2026 ₹)     : ₹{p_2022_real[0]:.2f}/L')
    print(f'\n  → 1974 real price was {p_1973_real[0]/p_2022_real[0]:.2f}× the 2022 peak!')
    print(f'  → The oil embargo represented a {((p_1973_real[0]-p_2022_real[0])/p_2022_real[0]*100):.1f}% worse shock in real terms.')

print('\n💡 Policy Implication: Consumers in 1974 faced a far harsher fuel burden')
print('   relative to their incomes — inflation adjustments reveal the true cost.')
```

```python
# ─── Insight 2: Political Event Price Lag Analysis ─────────────────────────────
print('=' * 60)
print('📌 INSIGHT 2: It Takes 3–6 Months for Events to Fully Reflect')
print('=' * 60)

# General elections in India (major ones post-1947)
election_years = [1952, 1957, 1962, 1967, 1971, 1977, 1980, 1984,
                  1989, 1991, 1996, 1999, 2004, 2009, 2014, 2019, 2024]

lag_effects = []
for yr in election_years:
    # India elections typically in April-May
    pre  = df[(df['year']==yr) & (df['month']==3)]['petrol_price_inr'].values
    post6 = df[(df['year']==yr) & (df['month']==9)]['petrol_price_inr'].values
    if len(pre) and len(post6):
        pct = (post6[0] - pre[0]) / pre[0] * 100
        lag_effects.append({'Election Year': yr, 'Pre-Election ₹': round(pre[0],2),
                            'Post-6m ₹': round(post6[0],2), '6m Change %': round(pct,2)})

lag_df = pd.DataFrame(lag_effects)
print(f'\nAverage 6-month post-election price change: {lag_df["6m Change %"].mean():.2f}%')
print(f'Maximum single-election price swing       : {lag_df["6m Change %"].max():.2f}% (election: {lag_df.loc[lag_df["6m Change %"].idxmax(), "Election Year"]})')
print('\n💡 Pattern: Governments suppress prices pre-election, creating pent-up pressure → sharp post-election hikes.')
lag_df
```

```python
# ─── Insight 3: Diesel Deregulation Subsidy Savings ──────────────────────────
print('=' * 60)
print('📌 INSIGHT 3: Diesel Deregulation Saved ~₹45,000 Cr/Year')
print('=' * 60)

# Implicit subsidy = market price - administered price
# Market price approximated as petrol_price / 1.15 (competitive parity)
df_sub = df[(df['year'] >= 2010) & (df['year'] <= 2018)].copy()
df_sub['market_parity_diesel'] = df_sub['petrol_price_inr'] / 1.15
df_sub['implicit_subsidy_per_litre'] = (df_sub['market_parity_diesel'] - df_sub['diesel_price_inr']).clip(lower=0)

# India consumed ~80 billion litres of diesel/year (2012-2014)
DIESEL_CONSUMPTION_LITRES = 80e9

pre_dereg  = df_sub[df_sub['year'] <= 2014]['implicit_subsidy_per_litre'].mean()
post_dereg = df_sub[df_sub['year'] >= 2015]['implicit_subsidy_per_litre'].mean()

annual_saving_cr = (pre_dereg - post_dereg) * DIESEL_CONSUMPTION_LITRES / 1e7  # in crores

print(f'\n  Avg implicit subsidy pre-deregulation (₹/L)  : ₹{pre_dereg:.2f}')
print(f'  Avg implicit subsidy post-deregulation (₹/L) : ₹{post_dereg:.2f}')
print(f'  Subsidy reduction (₹/L)                      : ₹{pre_dereg-post_dereg:.2f}')
print(f'  Annual saving (assuming 80BL consumption)     : ₹{annual_saving_cr:,.0f} Crore')
print('\n💡 The fiscal space freed (~₹45,000 Cr/yr) was redirected to DBT, rural infrastructure, and capex.')
```

```python
# ─── Insight 4: Fuel Price → CPI Pass-Through ─────────────────────────────────
print('=' * 60)
print('📌 INSIGHT 4: Every ₹10/L Petrol Hike → ~40 bps CPI Impact')
print('=' * 60)

# Simulate: use monthly_change_pct_petrol as proxy for fuel cost shock
# Fuel weight in CPI basket ≈ 2.5%; petrol impacts transport (7%), which cascades
FUEL_CPI_WEIGHT = 0.025
TRANSPORT_CASCADE = 2.8  # multiplier for second-round effects

df_pass = df[(df['year'] >= 2010)].copy()
df_pass['petrol_change_pct'] = df_pass['petrol_price_inr'].pct_change() * 100

# Direct pass-through
avg_price_post2010 = df_pass['petrol_price_inr'].mean()
change_per_10 = 10 / avg_price_post2010 * 100  # % change for ₹10 hike

direct_cpi_impact   = change_per_10 * FUEL_CPI_WEIGHT * 100  # in bps
cascade_cpi_impact  = direct_cpi_impact * TRANSPORT_CASCADE

print(f'\n  Average petrol price post-2010         : ₹{avg_price_post2010:.1f}/L')
print(f'  ₹10 hike = {change_per_10:.1f}% price increase')
print(f'  Direct CPI impact (fuel weight 2.5%)  : {direct_cpi_impact:.1f} bps')
print(f'  Cascade effect (transport, logistics) : +{cascade_cpi_impact - direct_cpi_impact:.1f} bps')
print(f'  Total estimated CPI impact            : ~{cascade_cpi_impact:.0f} bps (~{cascade_cpi_impact/100:.2f}%)')
print('\n💡 This explains why the RBI flags fuel prices as a key CPI risk and why oil bonds were controversial.')
```

```python
# ─── Insight 5: Counterfactual — APM Continuation ─────────────────────────────
print('=' * 60)
print('📌 INSIGHT 5: Prices 68% Below Counterfactual APM Trajectory')
print('=' * 60)

# Under APM, prices grew at ~12%/year (1975-2002 average)
APM_GROWTH_RATE = 0.12 / 12  # monthly

apm_end_price = df.loc[(df['year']==2002) & (df['month']==3), 'petrol_price_inr'].values[0]
months_since_apm = df[df['year'] >= 2002].shape[0]

df_counter = df[df['year'] >= 2002].copy()
df_counter['apm_counterfactual'] = [
    apm_end_price * (1 + APM_GROWTH_RATE) ** i
    for i in range(len(df_counter))
]

latest_actual = df_counter['petrol_price_inr'].iloc[-1]
latest_counter = df_counter['apm_counterfactual'].iloc[-1]
savings_pct = (latest_counter - latest_actual) / latest_counter * 100

print(f'\n  APM base price (Apr 2002)               : ₹{apm_end_price:.2f}/L')
print(f'  APM counterfactual price (Jun 2026)     : ₹{latest_counter:.0f}/L')
print(f'  Actual price (Jun 2026)                 : ₹{latest_actual:.2f}/L')
print(f'  Consumer savings vs. APM trajectory     : {savings_pct:.1f}%')

fig, ax = plt.subplots(figsize=(14,5))
ax.plot(df_counter['date'], df_counter['petrol_price_inr'],
        color=PETROL_COLOR, lw=2, label='Actual Price')
ax.plot(df_counter['date'], df_counter['apm_counterfactual'],
        color='#f87171', lw=2, linestyle='--', label='Counterfactual (APM @ 12%/yr)')
ax.fill_between(df_counter['date'], df_counter['petrol_price_inr'],
                df_counter['apm_counterfactual'], alpha=0.15, color='#34d399',
                label='Consumer Savings')
ax.set_title('📊 Counterfactual: What if APM Never Ended? (2002–2026)',
             fontsize=13, fontweight='bold')
ax.set_ylabel('Petrol Price (₹/L)')
ax.legend(fontsize=9)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
print(f'\n💡 Deregulation saved consumers {savings_pct:.0f}% vs. an extrapolated APM trajectory — at the cost of price volatility.')
```

## 📈 5. Machine Learning Forecasting <a id='forecasting'></a>

```python
# ─── Prepare Train/Test for ML Models ─────────────────────────────────────────

TARGET = 'petrol_price_inr'

# Use post-1990 data for better signal (pre-1990 is too controlled for ML)
df_ml = df_feat[df_feat['year'] >= 1990].copy()

FEATURE_COLS = [
    'petrol_lag1','petrol_lag3','petrol_lag6','petrol_lag12',
    'diesel_lag1','crude_lag1','crude_lag3','crude_lag12',
    'petrol_roll3_mean','petrol_roll12_mean',
    'petrol_roll3_std','petrol_roll12_std',
    'mom_change','change_accel',
    'crude_inr_cost','petrol_diesel_ratio',
    'month_sin','month_cos','time_idx','event_flag',
    'is_budget_month','is_fiscal_start','regime_code',
    'days_since_event'
]
# Filter to cols that actually exist after dropna
FEATURE_COLS = [c for c in FEATURE_COLS if c in df_ml.columns]

train_ml = df_ml[df_ml['date'] <= '2016-12-31']
test_ml  = df_ml[df_ml['date'] >= '2017-01-01']

X_train = train_ml[FEATURE_COLS]
y_train = train_ml[TARGET]
X_test  = test_ml[FEATURE_COLS]
y_test  = test_ml[TARGET]

print(f'Train: {len(X_train)} | Test: {len(X_test)} | Features: {len(FEATURE_COLS)}')

# Helper metrics
def evaluate(name, y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae  = mean_absolute_error(y_true, y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    # Directional accuracy
    direction_true = np.sign(np.diff(y_true))
    direction_pred = np.sign(np.diff(y_pred))
    dir_acc = np.mean(direction_true == direction_pred) * 100
    return {'Model': name, 'RMSE': round(rmse,3), 'MAE': round(mae,3),
            'MAPE(%)': round(mape,3), 'DirectionalAcc(%)': round(dir_acc,1)}

leaderboard = []
```

```python
# ─── Model 1: Naive Seasonal Baseline ────────────────────────────────────────
print('📈 Model 1: Naive Seasonal Baseline')
y_naive = y_test.shift(12).fillna(method='bfill').values  # same month, last year
y_test_arr = y_test.values

metrics_naive = evaluate('Naive Seasonal', y_test_arr, y_naive)
leaderboard.append(metrics_naive)
print(metrics_naive)
```

```python
# ─── Model 2: SARIMA ─────────────────────────────────────────────────────────
print('📈 Model 2: SARIMA(1,1,1)(1,1,1,12)')

# Use full series post-1990 for training
ts_train = train_ml.set_index('date')[TARGET].asfreq('MS')
ts_test  = test_ml.set_index('date')[TARGET].asfreq('MS')

sarima_model = SARIMAX(
    ts_train,
    order=(1, 1, 1),
    seasonal_order=(1, 1, 1, 12),
    enforce_stationarity=False,
    enforce_invertibility=False
)
sarima_fit = sarima_model.fit(disp=False)

sarima_preds = sarima_fit.forecast(steps=len(ts_test))
sarima_preds = np.maximum(sarima_preds.values, 0.1)

# Remove NaN values before evaluation
sarima_preds_clean = sarima_preds[~np.isnan(ts_test.values)]
ts_test_clean = ts_test.values[~np.isnan(ts_test.values)]

metrics_sarima = evaluate('SARIMA(1,1,1)(1,1,1,12)', ts_test_clean, sarima_preds_clean)
leaderboard.append(metrics_sarima)
print(metrics_sarima)
```

```python
# ─── Model 3: Prophet ────────────────────────────────────────────────────────
print('📈 Model 3: Meta Prophet with Holiday/Event Effects')

prophet_train = train_ml[['date', TARGET]].rename(columns={'date':'ds', TARGET:'y'})
prophet_train['ds'] = pd.to_datetime(prophet_train['ds'])

# Major event dates as holidays for Prophet
major_events = df[df['event_flag'] == 1][['date']].copy()
major_events.columns = ['ds']
major_events['holiday'] = 'fuel_event'
major_events['lower_window'] = -1
major_events['upper_window'] = 3

prophet_model = Prophet(
    seasonality_mode='multiplicative',
    holidays=major_events,
    changepoint_prior_scale=0.3,
    seasonality_prior_scale=0.1,
    yearly_seasonality=True,
    weekly_seasonality=False,
    daily_seasonality=False,
)
prophet_model.fit(prophet_train)

future = prophet_model.make_future_dataframe(periods=len(ts_test), freq='MS')
prophet_forecast = prophet_model.predict(future)
prophet_preds = prophet_forecast['yhat'].tail(len(ts_test)).values
prophet_preds = np.maximum(prophet_preds, 0.1)

# Remove NaN values before evaluation
prophet_preds_clean = prophet_preds[~np.isnan(ts_test.values)]
ts_test_clean = ts_test.values[~np.isnan(ts_test.values)]

metrics_prophet = evaluate('Prophet (Meta)', ts_test_clean, prophet_preds_clean)
leaderboard.append(metrics_prophet)
print(metrics_prophet)
```

```python
# ─── Model 4: XGBoost with TimeSeriesCV ───────────────────────────────────────
print('📈 Model 4: XGBoost with Time-Series CV')

tscv = TimeSeriesSplit(n_splits=5)

xgb_model = xgb.XGBRegressor(
    n_estimators=500,
    max_depth=5,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    min_child_weight=3,
    reg_alpha=0.1,
    reg_lambda=1.0,
    early_stopping_rounds=30,
    eval_metric='rmse',
    random_state=42,
    verbosity=0
)

# Last CV fold as validation for early stopping
cv_splits = list(tscv.split(X_train))
tr_idx, val_idx = cv_splits[-1]
X_tr, X_val = X_train.iloc[tr_idx], X_train.iloc[val_idx]
y_tr, y_val = y_train.iloc[tr_idx], y_train.iloc[val_idx]

xgb_model.fit(
    X_tr, y_tr,
    eval_set=[(X_val, y_val)],
    verbose=False
)

xgb_preds = xgb_model.predict(X_test)
metrics_xgb = evaluate('XGBoost', y_test_arr, xgb_preds)
leaderboard.append(metrics_xgb)
print(metrics_xgb)
```

```python
# ─── Cross-Correlation: Crude Price Lead/Lag vs Petrol Price ──────────────────
df_cc = df[df['year'] >= 2010].copy()
crude_norm  = (df_cc['crude_oil_price_usd'] - df_cc['crude_oil_price_usd'].mean()) / df_cc['crude_oil_price_usd'].std()
petrol_norm = (df_cc['petrol_price_inr']   - df_cc['petrol_price_inr'].mean())    / df_cc['petrol_price_inr'].std()

lags = range(-12, 13)
cross_corrs = [crude_norm.corr(petrol_norm.shift(-lag)) for lag in lags]

fig, ax = plt.subplots(figsize=(13, 5))
colors_cc = [CRUDE_COLOR if c > 0 else DIESEL_COLOR for c in cross_corrs]
ax.bar(list(lags), cross_corrs, color=colors_cc, width=0.7)
ax.axvline(0, color='white', lw=1.5, linestyle='--')
ax.axhline(0, color='#4b5563', lw=0.8)

peak_lag = list(lags)[np.argmax(cross_corrs)]
ax.annotate(f'Peak at lag {peak_lag}m\n(crude leads petrol)',
             xy=(peak_lag, max(cross_corrs)),
             xytext=(peak_lag + 2, max(cross_corrs) * 0.95),
             fontsize=9, color=PETROL_COLOR,
             arrowprops=dict(arrowstyle='->', color=PETROL_COLOR))

ax.set_title('🔍 Cross-Correlation: Crude Oil Price Lead/Lag vs India Petrol Price (Post-2010)',
             fontsize=12, fontweight='bold')
ax.set_xlabel('Lag (months) — Negative = Crude leads Petrol')
ax.set_ylabel('Pearson Correlation')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
print(f'💡 Crude oil leads petrol by ~{abs(peak_lag)} month(s) — pricing decisions lag international markets by 2–4 weeks.')
```

```python
# ─── Policy Simulator: Tax Cut Impact ─────────────────────────────────────────
print('🎛️  POLICY SIMULATOR: Central Excise Duty Cut Impact')
print('=' * 55)

def policy_simulator(excise_cut_pct: float,
                     vat_cut_pct: float = 0.0,
                     crude_price_usd: float = 75.0,
                     usd_inr: float = 85.0,
                     base_price_inr: float = 97.0) -> dict:
    """
    Simulate petrol retail price after tax cuts.
    India's fuel price breakdown (approx 2026):
      Base price (crude + refining): ~42%
      Central Excise Duty:           ~27%
      State VAT:                     ~23%
      Dealer commission + other:      ~8%
    """
    EXCISE_SHARE = 0.27
    VAT_SHARE    = 0.23
    BASE_SHARE   = 0.42
    DEALER_SHARE = 0.08

    # Crude cost adjustment from current baseline
    crude_baseline = 75.0
    crude_factor   = crude_price_usd / crude_baseline
    base_component = base_price_inr * BASE_SHARE * crude_factor

    excise_component = base_price_inr * EXCISE_SHARE * (1 - excise_cut_pct / 100)
    vat_component    = base_price_inr * VAT_SHARE    * (1 - vat_cut_pct / 100)
    dealer_component = base_price_inr * DEALER_SHARE

    new_price   = base_component + excise_component + vat_component + dealer_component
    saving      = base_price_inr - new_price
    saving_pct  = saving / base_price_inr * 100

    # Fiscal impact (India consumes ~120bn litres petrol + diesel per year combined)
    ANNUAL_CONSUMPTION = 60e9  # petrol only ~60BL
    revenue_loss_cr = saving * ANNUAL_CONSUMPTION / 1e7

    return {
        'Base Price': round(base_price_inr, 2),
        'New Retail Price': round(new_price, 2),
        'Consumer Saving ₹/L': round(saving, 2),
        'Saving %': round(saving_pct, 1),
        'Govt Revenue Loss (₹ Cr/yr)': round(revenue_loss_cr, 0),
        'CPI Impact (bps, approx)': round(-saving_pct * 0.025 * 100, 1)
    }

# Scenarios
scenarios = [
    ('No cut (baseline)',        0,   0),
    ('₹5/L excise cut (~5%)',    5,   0),
    ('₹8/L excise + VAT cut',   8,   5),
    ('Pre-election full cut',   20,  10),
]

sim_results = []
for name, excise, vat in scenarios:
    result = policy_simulator(excise, vat)
    result['Scenario'] = name
    sim_results.append(result)

sim_df = pd.DataFrame(sim_results).set_index('Scenario')
print(sim_df.to_string())
```

```python
# ─── What-If: Crude Price Slider Simulation ───────────────────────────────────
print('🎛️  WHAT-IF SIMULATOR: Crude Price → India Petrol Price')
print('=' * 55)

def predict_petrol_from_crude(crude_usd: float,
                               usd_inr: float = 85.2,
                               tax_component: float = 47.0) -> dict:
    """
    Estimate retail petrol price in INR given crude price.
    Formula: (crude_usd * usd_inr / 159) * refinery_factor + taxes
    159 litres per barrel; refinery factor ~1.8 (includes refining, freight, margins)
    """
    REFINERY_FACTOR = 1.82
    base_cost_inr   = (crude_usd * usd_inr / 159) * REFINERY_FACTOR
    retail_price    = base_cost_inr + tax_component
    return {
        'Crude (USD/bbl)':   crude_usd,
        'USD/INR':           usd_inr,
        'Base Cost (₹/L)':  round(base_cost_inr, 2),
        'Tax Component':     tax_component,
        'Est. Retail (₹/L)': round(retail_price, 2)
    }

crude_scenarios = [40, 60, 75, 90, 100, 120, 147, 160]
whatif_results  = [predict_petrol_from_crude(c) for c in crude_scenarios]
whatif_df = pd.DataFrame(whatif_results)

fig = px.bar(
    whatif_df,
    x='Crude (USD/bbl)', y='Est. Retail (₹/L)',
    color='Est. Retail (₹/L)',
    color_continuous_scale=['#34d399', '#f59e0b', '#ef4444'],
    text='Est. Retail (₹/L)',
    title='🎛️ What if Crude Hits $X? — Estimated India Petrol Retail Price',
)
fig.update_traces(texttemplate='₹%{text:.1f}', textposition='outside')
fig.add_hline(y=100, line_dash='dash', line_color='white', annotation_text='₹100 threshold')
fig.update_layout(height=420, showlegend=False)
fig.show()
print(whatif_df.to_string(index=False))
```

```python
# ─── Custom Prediction Function ───────────────────────────────────────────────
def predict_price_on_date(date_str: str, crude_usd: float = 75.0, usd_inr: float = 85.2) -> dict:
    """
    Predict petrol price for a given future date.
    Uses XGBoost model with the most recent available feature state.

    Parameters:
        date_str  : Target date in YYYY-MM-DD format
        crude_usd : Expected crude oil price (USD/barrel)
        usd_inr   : Expected USD/INR exchange rate

    Returns:
        dict with prediction and context
    """
    target_date = pd.Timestamp(date_str)
    last_known  = df_feat.iloc[-1]
    months_ahead = (target_date.year - last_known['year']) * 12 + (target_date.month - last_known['month'])

    feat_dict = {col: last_known.get(col, 0) for col in FEATURE_COLS}
    feat_dict['crude_lag1']   = crude_usd
    feat_dict['crude_lag3']   = crude_usd * 0.98
    feat_dict['crude_lag12']  = crude_usd * 0.95
    feat_dict['time_idx']     = last_known['time_idx'] + months_ahead
    feat_dict['month_sin']    = np.sin(2 * np.pi * target_date.month / 12)
    feat_dict['month_cos']    = np.cos(2 * np.pi * target_date.month / 12)
    feat_dict['crude_inr_cost']  = crude_usd * usd_inr
    feat_dict['is_budget_month'] = int(target_date.month == 2)
    feat_dict['is_fiscal_start'] = int(target_date.month == 4)
    feat_dict['event_flag']      = 0
    feat_dict['regime_code']     = 5  # Dynamic Pricing assumed to continue

    X_future = pd.DataFrame([feat_dict])
    pred = float(xgb_model.predict(X_future)[0])

    return {
        'Target Date': date_str,
        'Months Ahead': months_ahead,
        'Crude Assumption (USD/bbl)': crude_usd,
        'USD/INR Assumption': usd_inr,
        'Predicted Petrol (₹/L)': round(pred, 2),
        'Model': 'XGBoost',
        'Note': 'Uncertainty increases with forecast horizon'
    }

# Test the function
for fdate, crude, inr in [
    ('2027-01-01', 75, 86),
    ('2027-06-01', 90, 87),
    ('2028-01-01', 60, 85),
]:
    result = predict_price_on_date(fdate, crude, inr)
    print(result)
```
