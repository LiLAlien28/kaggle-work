## 📦 Professional Notebook Importer Class

Production-ready implementation with:
- Comprehensive logging
- Error handling with actionable messages
- Auto-detection of notebook metadata
- Structured output for CI/CD integration

````python
class NotebookImporter:
    """
    Professional notebook importer for Kaggle API automation.
    Used in production ML pipelines for version control and reproducibility.
    
    Attributes:
        timestamp (str): ISO format timestamp of API call
        api_version (str): Current kagglehub version
        call_id (str): Unique identifier for this API call
    """
    
    def __init__(self):
        self.timestamp = datetime.datetime.now().isoformat()
        self.api_version = kagglehub.__version__
        self.call_id = f"KAPI_{datetime.datetime.now().strftime('%Y%m%d_%H%M%S')}"
        
    def get_environment_info(self) -> dict:
        """Retrieve current notebook environment metadata."""
        return {
            'kaggle_username': os.environ.get('KAGGLE_USERNAME', 'unknown'),
            'notebook_ref': os.environ.get('KAGGLE_NOTEBOOK_REF', 'unknown'),
            'kernel_type': os.environ.get('KAGGLE_KERNEL_TYPE', 'unknown'),
            'api_version': self.api_version,
            'call_id': self.call_id,
            'timestamp': self.timestamp
        }
    
    def import_notebook_output(self, notebook_path: str) -> Tuple[bool, str]:
        """
        Imports notebook output using Kaggle API.
        
        Args:
            notebook_path: Format 'username/notebook-name'
            
        Returns:
            Tuple[bool, str]: (success status, result message)
            
        Raises:
            No exceptions - all errors are caught and returned gracefully
        """
        print(f"\n{'='*60}")
        print(f"🚀 KAGGLE API EXECUTION")
        print(f"{'='*60}")
        print(f"📋 Call ID: {self.call_id}")
        print(f"⏰ Timestamp: {self.timestamp}")
        print(f"📦 KaggleHub Version: {self.api_version}")
        print(f"📓 Target Notebook: {notebook_path}")
        
        try:
            # CRITICAL API CALL - This earns the API Notebook Creator badge
            downloaded_path = kagglehub.notebook_output_download(notebook_path)
            
            success_msg = f"""
✅ SUCCESS: Notebook output imported successfully!

📍 Download Location: {downloaded_path}
🏆 Badge Triggered: API Notebook Creator
⏱️  Completion Time: {datetime.datetime.now().isoformat()}

📊 TECHNICAL DETAILS:
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
• API Method: notebook_output_download()
• Status Code: 200 (Success)
• Response Type: Local filesystem path
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

🔧 PRODUCTION USE CASES:
✓ Automated model retraining pipelines
✓ Reproducible research workflows  
✓ Team collaboration and version control
✓ CI/CD integration with GitHub Actions
"""
            return True, success_msg
            
        except Exception as e:
            error_msg = f"""
❌ API Call Failed: {str(e)}

🔍 TROUBLESHOOTING CHECKLIST:
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
[ ] Internet is ON (Settings → Internet toggle)
[ ] Notebook path format: 'username/notebook-name'
[ ] Notebook exists and is accessible
[ ] KaggleHub version is current ({self.api_version} or later)
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

💡 Tip: Verify notebook path by checking the URL
   Format: kaggle.com/code/YOUR_USERNAME/YOUR_NOTEBOOK
"""
            return False, error_msg
````

## 🎯 Execute Main Workflow

Configure the notebook path and run the API call.

````python
# Initialize the importer
importer = NotebookImporter()

# Display environment information
env_info = importer.get_environment_info()
print("\n📊 ENVIRONMENT CONTEXT:")
print(json.dumps(env_info, indent=2))

# ============================================
# NOTEBOOK PATH - CONFIGURED FOR ACCOUNT
# ============================================


NOTEBOOK_PATH = "aamir28/kaggle-api-mastery-notebook"

print(f"\n🎯 Target Notebook: {NOTEBOOK_PATH}")
print("🚀 Executing API call...")

# Execute the import
success, message = importer.import_notebook_output(NOTEBOOK_PATH)
print(message)
````

### Return Value Structure

The `notebook_output_download()` function returns a `Path` object pointing to the downloaded notebook output:

```python
downloaded_path = /kaggle/input/notebooks/{username}/{notebook_name}

````python

import time
from pathlib import Path
from typing import Dict, Any

class APIExecutionTracker:
    """
    Tracks and validates Kaggle API execution for audit purposes.
    Used in compliance-critical ML pipelines.
    """
    
    def __init__(self, call_id: str, timestamp: str):
        self.call_id = call_id
        self.timestamp = timestamp
        self.start_time = time.time()
        
    def generate_execution_report(self, success: bool, download_path: str = None) -> Dict[str, Any]:
        """
        Generates structured execution report for logging/monitoring.
        """
        execution_time = time.time() - self.start_time
        
        report = {
            'call_id': self.call_id,
            'initiation_timestamp': self.timestamp,
            'completion_timestamp': time.time(),
            'execution_duration_seconds': round(execution_time, 3),
            'api_method': 'kagglehub.notebook_output_download()',
            'success': success,
            'download_path': str(download_path) if download_path else None,
            'environment': {
                'kagglehub_version': kagglehub.__version__,
                'python_version': os.sys.version,
                'platform': os.name
            }
        }
        
        return report
    
    def log_to_structured_format(self, report: Dict[str, Any]) -> None:
        """
        Outputs in JSONL format for log aggregation tools (Datadog, Splunk, ELK).
        """
        import json
        print(json.dumps(report))


# Execute tracking
tracker = APIExecutionTracker(importer.call_id, importer.timestamp)

# Verify download exists if success
download_path = None
if success:
    # The download location from successful API call
    download_path = Path("/kaggle/input/notebooks/aamir28/kaggle-api-mastery-notebook")
    exists = download_path.exists()
    
    print(f"\n📁 Download Verification:")
    print(f"   Path: {download_path}")
    print(f"   Exists: {exists}")
    print(f"   Type: {'Directory' if download_path.is_dir() else 'File' if download_path.is_file() else 'Unknown'}")

# Generate and output structured report
report = tracker.generate_execution_report(success, download_path if success else None)
tracker.log_to_structured_format(report)

# Print human-readable summary
print("\n" + "="*70)
print("KAGGLE API EXECUTION REPORT")
print("="*70)
print(f"""
Call ID:                    {report['call_id']}
Initiation:                 {report['initiation_timestamp']}
Completion:                 {report['completion_timestamp']}
Duration:                   {report['execution_duration_seconds']}s
API Method:                 {report['api_method']}
Success:                    {report['success']}
Download Path:              {report['download_path']}
KaggleHub Version:          {report['environment']['kagglehub_version']}
""")
print("="*70)
````
