## 3. Feature Engineering <a id='3'></a>


```python
# Derived features
df['is_placed']          = (df['placement_status'] == 'Placed').astype(int)
df['CGPA_band']          = pd.cut(df['cgpa'], bins=[4,6,7,8,9,10],
                                   labels=['<6','6-7','7-8','8-9','9+'])
df['internship_band']    = pd.cut(df['internships_completed'],
                                   bins=[-1,0,2,4,6],
                                   labels=['0','1-2','3-4','5-6'])
df['tech_score']         = (df['DSA_problems_solved']/df['DSA_problems_solved'].max()*40
                             + df['GitHub_repos']/df['GitHub_repos'].max()*30
                             + df['AI_ML_projects']/df['AI_ML_projects'].max()*30)
df['wellbeing_score']    = (100 - df['stress_level'] + 100 - df['burnout_score']
                             + df['sleep_hours']*6) / 3
df['ai_readiness']       = (df['AI_tool_usage_frequency'] + df['prompt_engineering_skill']
                             + df['adaptability_score'] - df['AI_fear_score']) / 4

print('Engineered features:')
print(df[['tech_score','wellbeing_score','ai_readiness','CGPA_band','internship_band']].describe().round(2))
```

## 5. Salary Distribution & Drivers <a id='5'></a>


```python
# CORRECTED VERSION - Salary Distribution & Key Drivers

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.gridspec as gridspec

# Define background and panel colors
BG = '#1a1a2e'
PANEL = '#2d2d44'

# Define company color palette
COMPANY_PAL = {
    'MNC': '#f59e0b',
    'Product-Based': '#38bdf8',
    'Startup': '#34d399',
    'Service-Based': '#94a3b8'
}

# Filter placed students
placed_df = df[df['is_placed'] == 1].copy()

fig = plt.figure(figsize=(18, 10), facecolor=BG)
gs = gridspec.GridSpec(2, 3, figure=fig, hspace=0.45, wspace=0.35)
fig.suptitle('Salary Distribution & Key Drivers', fontsize=17,
             fontweight='bold', color='white', y=0.98)

# 1. Overall histogram
ax = fig.add_subplot(gs[0, :])
ax.set_facecolor(PANEL)
ax.tick_params(colors='white')
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')
ax.title.set_color('white')

# Create histogram
n, bins, patches = ax.hist(placed_df['salary_lpa'].dropna(), bins=50,
                            color='#38bdf8', edgecolor='none', alpha=0.85, zorder=3)

# Gradient colouring by bin centre
for patch, left in zip(patches, bins[:-1]):
    t = (left - bins[0]) / (bins[-1] - bins[0])
    patch.set_facecolor(plt.cm.cool(t))
    patch.set_alpha(0.85)

# Add mean and median lines
mean_sal = placed_df['salary_lpa'].mean()
med_sal = placed_df['salary_lpa'].median()
ax.axvline(mean_sal, color='#f59e0b', lw=2, ls='--',
           label=f'Mean: ₹{mean_sal:.2f}L')
ax.axvline(med_sal, color='#f472b6', lw=2, ls=':',
           label=f'Median: ₹{med_sal:.2f}L')
ax.set_xlabel('Salary (LPA)', fontsize=12)
ax.set_ylabel('Count', fontsize=12)
ax.set_title('Overall Salary Distribution — Placed Students', fontsize=13, fontweight='bold')
ax.legend(fontsize=11, facecolor=PANEL, edgecolor='white')
ax.grid(axis='y', zorder=0, alpha=0.3)
ax.set_axisbelow(True)

# 2. By Company Type (violin)
ax2 = fig.add_subplot(gs[1, 0])
ax2.set_facecolor(PANEL)
ax2.tick_params(colors='white')
ax2.xaxis.label.set_color('white')
ax2.yaxis.label.set_color('white')
ax2.title.set_color('white')

ctypes = ['MNC', 'Product-Based', 'Startup', 'Service-Based']
xtick_labels = ['MNC', 'Product', 'Startup', 'Service']

for i, ct in enumerate(ctypes):
    data = placed_df[placed_df['company_type'] == ct]['salary_lpa'].dropna()
    if len(data) > 0:  # Check if data exists
        vp = ax2.violinplot(data, positions=[i], widths=0.6,
                            showmedians=True, showextrema=False)
        for pc in vp['bodies']: 
            pc.set_facecolor(COMPANY_PAL.get(ct, '#94a3b8'))
            pc.set_alpha(0.5)
        vp['cmedians'].set_color('white')
        vp['cmedians'].set_linewidth(2)
        
        # Add jittered scatter points (limit to 500 for performance)
        sample_size = min(len(data), 500)
        jitter = np.random.uniform(-0.1, 0.1, sample_size)
        ax2.scatter(i + jitter, data.sample(sample_size, random_state=42),
                   s=4, alpha=0.25, color=COMPANY_PAL.get(ct, '#94a3b8'), zorder=4)

ax2.set_xticks(range(4))
ax2.set_xticklabels(xtick_labels, fontsize=9)
ax2.set_ylabel('Salary (LPA)')
ax2.set_title('By Company Type', fontweight='bold')
ax2.grid(axis='y', zorder=0, alpha=0.3)
ax2.set_axisbelow(True)

# 3. By Internship count (step line)
ax3 = fig.add_subplot(gs[1, 1])
ax3.set_facecolor(PANEL)
ax3.tick_params(colors='white')
ax3.xaxis.label.set_color('white')
ax3.yaxis.label.set_color('white')
ax3.title.set_color('white')

# Check column name (assuming 'internships' or 'internships_completed')
intern_col = 'internships_completed' if 'internships_completed' in placed_df.columns else 'internships'
intern_means = placed_df.groupby(intern_col)['salary_lpa'].mean()
intern_counts = placed_df.groupby(intern_col)['salary_lpa'].count()

ax3.plot(intern_means.index, intern_means.values, color='#34d399',
         linewidth=2.5, marker='o', markersize=8, zorder=5)
ax3.fill_between(intern_means.index, intern_means.values,
                  alpha=0.15, color='#34d399')

# Add labels with vertical offset
max_y = intern_means.max()
offset = max_y * 0.02
for x, y in zip(intern_means.index, intern_means.values):
    ax3.text(x, y + offset, f'₹{y:.1f}L', ha='center', fontsize=9,
             color='#34d399', fontweight='bold')

ax3.set_xlabel('Internships Completed')
ax3.set_ylabel('Avg Salary (LPA)')
ax3.set_title('Salary vs Internships', fontweight='bold')
ax3.grid(True, zorder=0, alpha=0.3)
ax3.set_axisbelow(True)

# 4. By College Tier (box)
ax4 = fig.add_subplot(gs[1, 2])
ax4.set_facecolor(PANEL)
ax4.tick_params(colors='white')
ax4.xaxis.label.set_color('white')
ax4.yaxis.label.set_color('white')
ax4.title.set_color('white')

tier_colors = ['#f59e0b', '#38bdf8', '#94a3b8']
tier_data = [placed_df[placed_df['college_tier'] == str(t)]['salary_lpa'].dropna() for t in [1, 2, 3]]

bp = ax4.boxplot(tier_data, patch_artist=True,
                  medianprops=dict(color='white', linewidth=2),
                  whiskerprops=dict(color='#64748b'),
                  capprops=dict(color='#64748b'),
                  flierprops=dict(marker='o', markerfacecolor='#64748b',
                                  markersize=3, alpha=0.3, linestyle='none'))

for patch, color in zip(bp['boxes'], tier_colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.55)

ax4.set_xticklabels(['Tier 1', 'Tier 2', 'Tier 3'], color='white')
ax4.set_ylabel('Salary (LPA)')
ax4.set_title('Salary by College Tier', fontweight='bold')
ax4.grid(axis='y', zorder=0, alpha=0.3)
ax4.set_axisbelow(True)

plt.tight_layout()
plt.savefig('03_salary_analysis.png', dpi=150, bbox_inches='tight', facecolor=BG)
plt.show()
```

## 13. Branch-wise Career Outcomes <a id='13'></a>


```python
branches = df['branch'].value_counts().index.tolist()
b_cols   = ['#38bdf8','#818cf8','#f472b6','#34d399','#fb923c','#f59e0b']

fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.patch.set_facecolor(BG)
fig.suptitle('Branch-wise Placement Rate · Avg Salary · Layoff Risk', fontsize=16,
             fontweight='bold', color='white', y=1.01)

# 1. Placement rate
ax = axes[0]
b_rate = [df[df['branch']==b]['is_placed'].mean()*100 for b in branches]
y_pos  = range(len(branches))
bars = ax.barh(y_pos, b_rate, color=b_cols, height=0.6, zorder=3, edgecolor='none')
ax.set_yticks(y_pos); ax.set_yticklabels(branches, fontsize=10)
ax.set_xlabel('Placement Rate (%)')
ax.set_title('Placement Rate by Branch', fontweight='bold')
ax.set_xlim(95, 100)
for bar, v in zip(bars, b_rate):
    ax.text(bar.get_width()+0.02, bar.get_y()+bar.get_height()/2,
            f'{v:.1f}%', va='center', fontsize=9, color='white')
ax.grid(axis='x', zorder=0); ax.set_axisbelow(True)

# 2. Avg salary
ax = axes[1]
b_sal = [placed_df[placed_df['branch']==b]['salary_lpa'].mean() for b in branches]
bars = ax.barh(y_pos, b_sal, color=b_cols, height=0.6, zorder=3, edgecolor='none')
ax.set_yticks(y_pos); ax.set_yticklabels(branches, fontsize=10)
ax.set_xlabel('Avg Salary (LPA)')
ax.set_title('Avg Salary by Branch', fontweight='bold')
for bar, v in zip(bars, b_sal):
    ax.text(bar.get_width()+0.05, bar.get_y()+bar.get_height()/2,
            f'₹{v:.2f}L', va='center', fontsize=9, color='white')
ax.grid(axis='x', zorder=0); ax.set_axisbelow(True)

# 3. Layoff risk
ax = axes[2]
b_risk = [placed_df[placed_df['branch']==b]['layoffs_risk_score'].mean() for b in branches]
bars = ax.barh(y_pos, b_risk, color=b_cols, height=0.6, zorder=3, edgecolor='none')
ax.set_yticks(y_pos); ax.set_yticklabels(branches, fontsize=10)
ax.set_xlabel('Avg Layoff Risk Score')
ax.set_title('Layoff Risk by Branch', fontweight='bold')
for bar, v in zip(bars, b_risk):
    ax.text(bar.get_width()+0.1, bar.get_y()+bar.get_height()/2,
            f'{v:.1f}', va='center', fontsize=9, color='white')
ax.grid(axis='x', zorder=0); ax.set_axisbelow(True)

plt.tight_layout()
plt.savefig('12_branch_analysis.png', dpi=150, bbox_inches='tight', facecolor=BG)
plt.show()
```
