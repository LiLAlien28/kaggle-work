## 1. Setup & Imports <a id='1'></a>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    roc_auc_score, classification_report, confusion_matrix,
    RocCurveDisplay, ConfusionMatrixDisplay
)

try:
    import lightgbm as lgb
    HAS_LGB = True
except ImportError:
    HAS_LGB = False
    print('LightGBM not found — falling back to GradientBoosting')

# ── aesthetics ──────────────────────────────────────────────────────────────
PALETTE  = ['#2ecc71', '#e74c3c']   # green=legit, red=fraud
sns.set_theme(style='whitegrid', font_scale=1.1)
plt.rcParams['figure.dpi'] = 120

SEED = 42
np.random.seed(SEED)
print('✅ Imports OK')
```

## 2. Data Loading & Overview <a id='2'></a>

```python
# ── load ────────────────────────────────────────────────────────────────────
df = pd.read_csv('/kaggle/input/datasets/lukhilaksh/strategic-voting-machine/Vote Ai.csv')   

print(f'Shape  : {df.shape}')
print(f'Target : Fraud_Label')
df.head(3)
```

```python
# ── basic info ───────────────────────────────────────────────────────────────
print('═'*55)
print(f" Rows          : {df.shape[0]:,}")
print(f" Columns       : {df.shape[1]}")
print(f" Numeric cols  : {df.select_dtypes('number').shape[1]}")
print(f" Categorical   : {df.select_dtypes('object').shape[1]}")
missing_pct = df.isnull().mean().mul(100).round(1)
print(f" Cols w/ NaN   : {(missing_pct>0).sum()}")
print('═'*55)

df.info(show_counts=True)
```

```python
# ── target distribution ──────────────────────────────────────────────────────
counts = df['Fraud_Label'].value_counts()
labels = ['Legit (0)', 'Fraud (1)']

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].bar(labels, counts.values, color=PALETTE, edgecolor='white', linewidth=1.4)
for i, v in enumerate(counts.values):
    axes[0].text(i, v + 120, f'{v:,}', ha='center', fontweight='bold')
axes[0].set_title('Class Distribution', fontweight='bold')
axes[0].set_ylabel('Count')

axes[1].pie(counts.values, labels=labels, colors=PALETTE,
            autopct='%1.1f%%', startangle=140,
            wedgeprops=dict(edgecolor='white', linewidth=2))
axes[1].set_title('Class Proportion', fontweight='bold')

plt.suptitle('Target Variable — Fraud_Label', fontsize=13, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()

print(f"Imbalance ratio  (legit:fraud) ≈  {counts[0]/counts[1]:.2f} : 1")
```

## 3. Exploratory Data Analysis <a id='3'></a>

```python
# ── missing values heatmap ───────────────────────────────────────────────────
miss = missing_pct[missing_pct > 0].sort_values(ascending=False)

fig, ax = plt.subplots(figsize=(10, 5))
bars = ax.barh(miss.index, miss.values,
               color=plt.cm.RdYlGn_r(miss.values / 100), edgecolor='none')
for bar, val in zip(bars, miss.values):
    ax.text(val + 0.3, bar.get_y() + bar.get_height()/2,
            f'{val:.1f}%', va='center', fontsize=9)
ax.set_xlabel('Missing %')
ax.set_title('Missing Values by Column', fontweight='bold')
ax.invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# ── risk category breakdown ───────────────────────────────────────────────────
risk_order = ['Low', 'Medium', 'High', 'Critical']
pivot = (
    df.groupby(['Risk_Category', 'Fraud_Label'])
    .size().unstack(fill_value=0)
    .reindex(risk_order)
)
pivot.columns = ['Legit', 'Fraud']

pivot.plot(kind='bar', color=PALETTE, edgecolor='white', linewidth=1.2,
           figsize=(8, 4), rot=0)
plt.title('Fraud Count by Risk Category', fontweight='bold')
plt.ylabel('Count')
plt.legend()
plt.tight_layout()
plt.show()

# fraud rate per category
pivot['Fraud Rate %'] = (pivot['Fraud'] / (pivot['Legit'] + pivot['Fraud']) * 100).round(1)
print(pivot)
```

```python
# ── correlation heatmap (numeric, top 20 by variance) ────────────────────────
num_df = df.select_dtypes('number').drop(columns=['Fraud_Label'], errors='ignore')
top20  = num_df.var().nlargest(20).index
corr   = num_df[top20].corr()

fig, ax = plt.subplots(figsize=(12, 10))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, cmap='coolwarm', center=0,
            annot=True, fmt='.2f', linewidths=.4,
            annot_kws={'size': 7}, ax=ax)
ax.set_title('Correlation Heatmap — Top 20 Numeric Features', fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# ── state-level fraud rate ────────────────────────────────────────────────────
state_fraud = (
    df.groupby('State')['Fraud_Label']
    .agg(['mean', 'count'])
    .query('count >= 50')
    .rename(columns={'mean': 'fraud_rate', 'count': 'n'})
    .sort_values('fraud_rate', ascending=False)
    .head(20)
)

fig, ax = plt.subplots(figsize=(11, 5))
bars = ax.barh(state_fraud.index, state_fraud['fraud_rate'] * 100,
               color='#e74c3c', alpha=0.8, edgecolor='none')
ax.axvline(df['Fraud_Label'].mean() * 100, color='#2ecc71',
           linestyle='--', linewidth=2, label=f'Overall avg ({df["Fraud_Label"].mean()*100:.1f}%)')
ax.set_xlabel('Fraud Rate (%)')
ax.set_title('Top 20 States by Fraud Rate (min 50 records)', fontweight='bold')
ax.legend()
ax.invert_yaxis()
plt.tight_layout()
plt.show()
```

## 4. Feature Engineering <a id='4'></a>

```python
# ── drop columns that are leakage / pure noise / identifiers ─────────────────
DROP_COLS = [
    'Election_ID',           # unique ID
    'Constituency_Code',     # high-cardinality ID
    'EVM_Device_ID',         # device identifier
    'Last_Updated_Time',     # timestamp (not parsed here)
    'Noise_Float_1',         # explicit noise
    'Noise_Float_2',
    'Noise_Category',
    'Fraud_Proxy_Duplicate', # near-duplicate of target
    'Election_Result',       # outcome leakage
    'Risk_Category',         # derived from target
]

df_eng = df.drop(columns=DROP_COLS, errors='ignore').copy()

# ── new features ─────────────────────────────────────────────────────────────
# composite risk score
df_eng['composite_risk'] = (
    df_eng['Fraud_Risk_Score'].fillna(0) +
    df_eng['Anomaly_Score'].fillna(0) +
    df_eng['Strategic_Manipulation_Score'].fillna(0)
) / 3

# vote efficiency (EVM vs total)
df_eng['evm_ratio'] = df_eng['EVM_Votes'] / (df_eng['Total_Votes'] + 1)

# assets-to-liability ratio (capped)
df_eng['asset_liability_ratio'] = (
    df_eng['Assets_Cr'] / (df_eng['Liabilities_Cr'] + 0.01)
).clip(0, 100)

# urban-rural balance
df_eng['urban_rural_gap'] = abs(df_eng['Urban_Voter_Ratio'] - df_eng['Rural_Voter_Ratio'])

print(f'Features after engineering: {df_eng.shape[1] - 1}')  # minus target
df_eng.head(2)
```

## 5. Preprocessing Pipeline <a id='5'></a>

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder

TARGET = 'Fraud_Label'
X = df_eng.drop(columns=[TARGET])
y = df_eng[TARGET]

num_cols = X.select_dtypes('number').columns.tolist()
cat_cols = X.select_dtypes('object').columns.tolist()

print(f'Numeric  : {len(num_cols)}')
print(f'Categorical : {cat_cols}')

num_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler',  StandardScaler())
])

cat_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
])

preprocessor = ColumnTransformer([
    ('num', num_pipe, num_cols),
    ('cat', cat_pipe, cat_cols)
])

print('\n✅ Preprocessor defined')
```

## 6. Modeling & Cross-Validation <a id='6'></a>

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)

# ── model zoo ─────────────────────────────────────────────────────────────────
models = {
    'Logistic Regression': Pipeline([
        ('pre', preprocessor),
        ('clf', LogisticRegression(max_iter=2000, C=0.5, solver='liblinear', random_state=SEED))
    ]),
    'Random Forest': Pipeline([
        ('pre', preprocessor),
        ('clf', RandomForestClassifier(n_estimators=200, max_depth=12,
                                       n_jobs=-1, random_state=SEED))
    ]),
}

if HAS_LGB:
    models['LightGBM'] = Pipeline([
        ('pre', preprocessor),
        ('clf', lgb.LGBMClassifier(
            n_estimators=500, learning_rate=0.05,
            num_leaves=63, min_child_samples=20,
            subsample=0.8, colsample_bytree=0.8,
            scale_pos_weight=counts[0]/counts[1],  # handle imbalance
            random_state=SEED, n_jobs=-1, verbose=-1
        ))
    ])
else:
    models['GradientBoosting'] = Pipeline([
        ('pre', preprocessor),
        ('clf', GradientBoostingClassifier(
            n_estimators=200, learning_rate=0.1,
            max_depth=5, random_state=SEED
        ))
    ])

print(f'Training {len(models)} model(s) with 5-fold CV…')
results = {}

for name, pipe in models.items():
    print(f'  Running {name}...')
    scores = cross_val_score(pipe, X, y, cv=cv, scoring='roc_auc', n_jobs=1)  # Changed to n_jobs=1 to avoid hangs
    results[name] = scores
    print(f'  {name:25s} — AUC: {scores.mean():.4f} ± {scores.std():.4f}')

print('\n✅ CV complete')
```

```python
# ── compare models visually ───────────────────────────────────────────────────
res_df = pd.DataFrame(results).melt(var_name='Model', value_name='AUC')

# Handle case where results might be empty
if res_df.empty or len(res_df['AUC'].dropna()) == 0:
    print("⚠️ No valid results to plot")
else:
    fig, ax = plt.subplots(figsize=(9, 5))
    
    # Determine y-axis limits dynamically
    y_min = max(0.7, res_df['AUC'].min() - 0.01)
    y_max = min(1.01, res_df['AUC'].max() + 0.01)
    
    sns.boxplot(data=res_df, x='Model', y='AUC', palette='Set2',
                width=0.45, linewidth=1.4, fliersize=4, ax=ax)
    sns.stripplot(data=res_df, x='Model', y='AUC', color='black',
                  alpha=0.6, jitter=True, size=5, ax=ax)
    
    ax.set_ylim(y_min, y_max)
    ax.set_title('5-Fold CV AUC — Model Comparison', fontweight='bold', fontsize=12)
    ax.set_xlabel('')
    ax.set_ylabel('AUC Score')
    ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
```

## 7. Evaluation & Insights <a id='7'></a>

```python
from sklearn.model_selection import train_test_split

# pick best model
best_name = max(results, key=lambda k: results[k].mean())
best_pipe  = models[best_name]
print(f'Best model: {best_name}')

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=SEED
)

best_pipe.fit(X_train, y_train)
y_prob = best_pipe.predict_proba(X_test)[:, 1]
y_pred = (y_prob >= 0.5).astype(int)

print(f'\nTest AUC : {roc_auc_score(y_test, y_prob):.4f}')
print('\nClassification Report:')
print(classification_report(y_test, y_pred, target_names=['Legit', 'Fraud']))
```

```python
# ── ROC + Confusion Matrix ────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

RocCurveDisplay.from_predictions(y_test, y_prob, ax=axes[0],
                                  name=best_name, color='#e74c3c')
axes[0].plot([0,1],[0,1], '--', color='grey')
axes[0].set_title('ROC Curve', fontweight='bold')

ConfusionMatrixDisplay.from_predictions(
    y_test, y_pred, display_labels=['Legit', 'Fraud'],
    colorbar=False, cmap='Reds', ax=axes[1]
)
axes[1].set_title('Confusion Matrix', fontweight='bold')

plt.suptitle(f'{best_name} — Test Set Performance', fontweight='bold', fontsize=13)
plt.tight_layout()
plt.show()
```

```python
# ── feature importance ────────────────────────────────────────────────────────
clf = best_pipe.named_steps['clf']
pre = best_pipe.named_steps['pre']

feature_names = (
    num_cols +
    cat_cols
)

if hasattr(clf, 'feature_importances_'):
    fi = pd.Series(clf.feature_importances_, index=feature_names)
    top_fi = fi.nlargest(20).sort_values()

    fig, ax = plt.subplots(figsize=(9, 7))
    colors = ['#e74c3c' if v > top_fi.median() else '#3498db' for v in top_fi]
    top_fi.plot(kind='barh', color=colors, ax=ax, edgecolor='none')
    ax.set_title('Top 20 Feature Importances', fontweight='bold')
    ax.set_xlabel('Importance')
    plt.tight_layout()
    plt.show()
else:
    print('Feature importances not available for this model.')
```

```python
# ── threshold analysis ────────────────────────────────────────────────────────
from sklearn.metrics import precision_score, recall_score, f1_score

thresholds = np.arange(0.1, 0.91, 0.05)
metrics = {'Threshold': [], 'Precision': [], 'Recall': [], 'F1': []}

for t in thresholds:
    preds = (y_prob >= t).astype(int)
    metrics['Threshold'].append(t)
    metrics['Precision'].append(precision_score(y_test, preds, zero_division=0))
    metrics['Recall'].append(recall_score(y_test, preds, zero_division=0))
    metrics['F1'].append(f1_score(y_test, preds, zero_division=0))

thresh_df = pd.DataFrame(metrics)

fig, ax = plt.subplots(figsize=(10, 4))
for col, color in zip(['Precision', 'Recall', 'F1'], ['#3498db', '#e74c3c', '#2ecc71']):
    ax.plot(thresh_df['Threshold'], thresh_df[col], label=col, color=color, lw=2)
ax.axvline(0.5, color='grey', linestyle='--', label='Default (0.5)')
ax.set_xlabel('Decision Threshold')
ax.set_ylabel('Score')
ax.set_title('Precision / Recall / F1 vs Threshold', fontweight='bold')
ax.legend()
plt.tight_layout()
plt.show()

best_thresh = thresh_df.loc[thresh_df['F1'].idxmax()]
print(f'Best threshold (max F1): {best_thresh["Threshold"]:.2f}')
print(f'  Precision: {best_thresh["Precision"]:.3f}')
print(f'  Recall   : {best_thresh["Recall"]:.3f}')
print(f'  F1       : {best_thresh["F1"]:.3f}')
```

## 8. Submission <a id='8'></a>

```python
# ── retrain on full data ──────────────────────────────────────────────────────
print('Retraining on full dataset…')
best_pipe.fit(X, y)

# ── If a test file exists, predict on it ─────────────────────────────────────
import os
test_path = '/kaggle/input/vote-ai/test.csv'

if os.path.exists(test_path):
    test_df = pd.read_csv(test_path)
    test_eng = test_df.drop(columns=DROP_COLS, errors='ignore').copy()

    # same feature engineering
    test_eng['composite_risk'] = (
        test_eng['Fraud_Risk_Score'].fillna(0) +
        test_eng['Anomaly_Score'].fillna(0) +
        test_eng['Strategic_Manipulation_Score'].fillna(0)
    ) / 3
    test_eng['evm_ratio'] = test_eng['EVM_Votes'] / (test_eng['Total_Votes'] + 1)
    test_eng['asset_liability_ratio'] = (
        test_eng['Assets_Cr'] / (test_eng['Liabilities_Cr'] + 0.01)
    ).clip(0, 100)
    test_eng['urban_rural_gap'] = abs(test_eng['Urban_Voter_Ratio'] - test_eng['Rural_Voter_Ratio'])

    X_test_final = test_eng[X.columns]
    preds = best_pipe.predict_proba(X_test_final)[:, 1]

    submission = pd.DataFrame({
        'Election_ID': test_df['Election_ID'],
        'Fraud_Label': preds
    })
    submission.to_csv('submission.csv', index=False)
    print('✅ submission.csv saved!')
    submission.head()
else:
    # Self-evaluation mode: predict on training data for demo
    print('No test.csv found — running self-evaluation mode.')
    final_probs = best_pipe.predict_proba(X)[:, 1]
    final_preds = (final_probs >= float(best_thresh['Threshold'])).astype(int)
    print(f'Optimal threshold   : {best_thresh["Threshold"]:.2f}')
    print(f'Train AUC (full fit): {roc_auc_score(y, final_probs):.4f}')
    print('\n✅ Model is ready for deployment or submission.')
```
