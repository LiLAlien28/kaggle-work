```python
# ========================================
# 2. DATA LOADING & INITIAL OVERVIEW
# ========================================

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print("🚀 Loading all dataset files...\n")

# Load all main files
cities = pd.read_parquet('/kaggle/input/datasets/aamir28/global-urban-mobility-and-smart-cities-dataset/cities_metadata.parquet')
traffic = pd.read_parquet('/kaggle/input/datasets/aamir28/global-urban-mobility-and-smart-cities-dataset/daily_traffic_metrics.parquet')
modes = pd.read_parquet('/kaggle/input/datasets/aamir28/global-urban-mobility-and-smart-cities-dataset/mobility_modes.parquet')
env = pd.read_parquet('/kaggle/input/datasets/aamir28/global-urban-mobility-and-smart-cities-dataset/environmental_factors.parquet')
incidents = pd.read_parquet('/kaggle/input/datasets/aamir28/global-urban-mobility-and-smart-cities-dataset/incidents_events.parquet')
agg = pd.read_parquet('/kaggle/input/datasets/aamir28/global-urban-mobility-and-smart-cities-dataset/aggregated_daily_indicators.parquet')

# Basic Information
print(f"✅ Successfully loaded {len(cities)} cities dataset")
print(f"   • Traffic records      : {len(traffic):,}")
print(f"   • Mobility modes rows  : {len(modes):,}")
print(f"   • Environmental records: {len(env):,}")
print(f"   • Incidents            : {len(incidents):,}")
print(f"   • Date Range           : {traffic['date'].min()} to {traffic['date'].max()}\n")

# City List
print("🌍 Cities Included:")
display(cities[['city_name', 'country', 'region', 'population_2025']].style.hide(axis="index"))

# Summary Table
summary = pd.DataFrame({
    'File': ['cities_metadata', 'daily_traffic_metrics', 'mobility_modes', 
             'environmental_factors', 'incidents_events', 'aggregated_daily_indicators'],
    'Rows': [len(cities), len(traffic), len(modes), len(env), len(incidents), len(agg)],
    'Columns': [cities.shape[1], traffic.shape[1], modes.shape[1], 
                env.shape[1], incidents.shape[1], agg.shape[1]],
    'Description': [
        'City metadata and coordinates',
        'Hourly traffic & congestion data',
        'Ride-hailing, transit & EV data',
        'Weather & air quality',
        'Incidents & disruptions',
        'Daily aggregated metrics'
    ]
})

print("\n📋 Dataset Files Summary:")
display(summary.style.set_properties(**{'text-align': 'left'}))
```

```python
# ========================================
# 3. EXPLORATORY DATA ANALYSIS - TRAFFIC PATTERNS
# ========================================

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Set style
plt.style.use('default')
sns.set_palette("husl")

print(" Creating Visualizations...\n")

# 1. Average Congestion by Hour of Day (Global)
hourly = traffic.groupby('hour_of_day')['congestion_level'].mean().reset_index()

fig1 = px.line(hourly, x='hour_of_day', y='congestion_level', 
               title=' Average Congestion Level by Hour of Day (All Cities)',
               labels={'hour_of_day': 'Hour of Day', 'congestion_level': 'Avg Congestion Level'},
               markers=True, template='plotly_white')

fig1.update_layout(height=500, title_font_size=20)
fig1.show()
```

```python
# 2. Congestion Comparison by City
city_congestion = traffic.groupby('city_id')['congestion_level'].mean().reset_index()
city_congestion = city_congestion.merge(cities[['city_id', 'city_name', 'region']], on='city_id')

fig2 = px.bar(city_congestion.sort_values('congestion_level', ascending=False), 
              x='city_name', y='congestion_level', color='region',
              title=' Average Congestion Level by City',
              labels={'congestion_level': 'Average Congestion Level', 'city_name': 'City'},
              template='plotly_white')

fig2.update_layout(height=550, xaxis_tickangle=-45)
fig2.show()
```

```python
# 3. Speed vs Congestion (Scatter)
fig3 = px.scatter(traffic.sample(15000, random_state=42), 
                  x='congestion_level', y='avg_speed_kmh',
                  color='city_id', opacity=0.6,
                  title='Speed vs Congestion Level (Sampled)',
                  labels={'congestion_level': 'Congestion Level', 
                          'avg_speed_kmh': 'Average Speed (km/h)'},
                  template='plotly_white')

fig3.show()
```

```python
# ========================================
# 4. WEATHER IMPACT & MULTIMODAL ANALYSIS
# ========================================

# Fix date types for all dataframes
for df in [traffic, env, modes]:
    date_col = [c for c in df.columns if 'date' in c.lower()][0]
    df[date_col] = pd.to_datetime(df[date_col])

# Merge traffic and environmental data
merged = pd.merge(traffic, env, on=['date', 'city_id'], how='left')

print("🌧️ Weather Impact on Traffic\n")

# 1. Weather vs Congestion
weather_impact = merged.groupby('weather_condition').agg(
    congestion=('congestion_level', 'mean'),
    speed=('avg_speed_kmh', 'mean'),
    occupancy=('road_occupancy_pct', 'mean')
).round(2).reset_index()

fig1 = px.bar(weather_impact, x='weather_condition', y='congestion',
              title='Impact of Weather on Traffic Congestion',
              color='congestion', template='plotly_white', text='congestion')
fig1.update_traces(textposition='outside')
fig1.update_layout(height=500)
fig1.show()

```

```python
# 2. Temperature vs Congestion
fig2 = px.scatter(merged.sample(20000, random_state=42), 
                  x='temperature_c', y='congestion_level',
                  color='weather_condition', opacity=0.6,
                  trendline="ols",
                  title='Temperature vs Congestion Level',
                  template='plotly_white')
fig2.show()
```

```python
# 3. Multimodal Mobility
print("🚕 Multimodal Mobility Analysis\n")

modes_summary = modes.merge(cities[['city_id', 'city_name', 'region']], on='city_id')

fig3 = px.scatter(modes_summary, x='ride_hailing_trips', y='public_transit_trips',
                  color='region', hover_name='city_name', size='ev_percentage',
                  title='Ride Hailing vs Public Transit Usage by City',
                  labels={'ride_hailing_trips': 'Ride Hailing Trips',
                          'public_transit_trips': 'Public Transit Trips'},
                  template='plotly_white')
fig3.show()
```

```python
# 4. EV Adoption Trend
ev_trend = modes.groupby(modes['date'].dt.year)['ev_percentage'].mean().reset_index()
ev_trend.columns = ['Year', 'Avg_EV_Percentage']

fig4 = px.line(ev_trend, x='Year', y='Avg_EV_Percentage', 
               title='EV Adoption Trend (2020-2026)',
               markers=True, template='plotly_white')
fig4.show()

print("✅ Weather & Multimodal visualizations completed!")
```

```python
# ========================================
# 5. INCIDENTS ANALYSIS & MARKET COMPARISON
# ========================================

import plotly.graph_objects as go
from plotly.subplots import make_subplots

# First, check actual column names
print("🔍 Dataset Column Names:")
print(f"   incidents: {list(incidents.columns)}")
print(f"   traffic: {list(traffic.columns)}")
print(f"   cities: {list(cities.columns)}")

# Fix date types dynamically
date_col_incidents = [col for col in incidents.columns if 'date' in col.lower() or 'time' in col.lower()][0] if any('date' in col.lower() or 'time' in col.lower() for col in incidents.columns) else None
date_col_traffic = [col for col in traffic.columns if 'date' in col.lower() or 'time' in col.lower()][0] if any('date' in col.lower() or 'time' in col.lower() for col in traffic.columns) else None

if date_col_incidents:
    incidents[date_col_incidents] = pd.to_datetime(incidents[date_col_incidents])
if date_col_traffic:
    traffic[date_col_traffic] = pd.to_datetime(traffic[date_col_traffic])

print(f"\n   Date columns found: incidents='{date_col_incidents}', traffic='{date_col_traffic}'")

# ========================================
# SECTION 1: INCIDENT TYPE DISTRIBUTION
# ========================================
print("\n🚨 Section 1: Incident Type Distribution")

type_col = [col for col in incidents.columns if 'type' in col.lower() or 'category' in col.lower() or 'event' in col.lower()]
type_col = type_col[0] if type_col else incidents.columns[1]  # fallback to 2nd column

incident_counts = incidents[type_col].value_counts().reset_index()
incident_counts.columns = ['Incident Type', 'Count']

fig1 = px.pie(incident_counts, values='Count', names='Incident Type',
              title='📊 Distribution of Incident Types Across All Cities',
              hole=0.45, template='plotly_white',
              color_discrete_sequence=px.colors.sequential.Blues_r)

fig1.update_traces(textinfo='percent+label', textfont_size=12, 
                   marker=dict(line=dict(color='white', width=2)))
fig1.update_layout(height=520, title_font_size=22, 
                   legend=dict(font=dict(size=13)))
fig1.show()
```

```python
# ========================================
# SECTION 2: INCIDENTS BY DAY OF WEEK
# ========================================
if date_col_incidents:
    print("\n📅 Section 2: Incidents by Day of Week")
    
    incidents['day_of_week'] = incidents[date_col_incidents].dt.day_name()
    day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    incident_by_day = incidents.groupby('day_of_week').size().reindex(day_order).reset_index()
    incident_by_day.columns = ['Day', 'Incident Count']
    
    fig2 = px.bar(incident_by_day, x='Day', y='Incident Count',
                  title='📅 Weekly Pattern of Incidents',
                  color='Incident Count', template='plotly_white',
                  color_continuous_scale='Blues',
                  text='Incident Count')
    
    fig2.update_traces(textposition='outside', textfont=dict(size=13, color='#1e3a8a'))
    fig2.update_layout(height=480, title_font_size=22,
                       xaxis_title="", yaxis_title="Number of Incidents")
    fig2.show()
```

```python
# ========================================
# SECTION 3: SEVERITY IMPACT ON TRAFFIC
# ========================================
print("\n⚠️ Section 3: Severity Impact Analysis")

severity_col = [col for col in incidents.columns if 'severity' in col.lower() or 'level' in col.lower() or 'impact' in col.lower()]
severity_col = severity_col[0] if severity_col else None

if severity_col and date_col_incidents and date_col_traffic:
    # Rename date columns to match for merge
    incidents_merge = incidents.copy()
    traffic_merge = traffic.copy()
    
    if date_col_incidents != date_col_traffic:
        incidents_merge.rename(columns={date_col_incidents: 'date'}, inplace=True)
        traffic_merge.rename(columns={date_col_traffic: 'date'}, inplace=True)
        merge_date_col = 'date'
    else:
        merge_date_col = date_col_incidents
    
    # Merge on city_id and date
    incident_traffic = pd.merge(
        incidents_merge, traffic_merge, 
        on=['city_id', merge_date_col],
        suffixes=('_inc', '_traf')
    )
    
    # Find congestion and speed columns
    cong_col = [col for col in incident_traffic.columns if 'congestion' in col.lower()]
    speed_col = [col for col in incident_traffic.columns if 'speed' in col.lower()]
    
    if cong_col and speed_col:
        cong_col, speed_col = cong_col[0], speed_col[0]
        
        severity_impact = incident_traffic.groupby(severity_col).agg({
            cong_col: 'mean',
            speed_col: 'mean'
        }).round(2).reset_index()
        
        severity_impact.columns = ['Severity', 'Avg Congestion', 'Avg Speed']
        
        fig3 = make_subplots(specs=[[{"secondary_y": True}]])
        
        fig3.add_trace(
            go.Bar(name='Avg Congestion Level', 
                   x=severity_impact['Severity'],
                   y=severity_impact['Avg Congestion'], 
                   marker=dict(color='#3b82f6', line=dict(color='#1e40af', width=1.5)),
                   opacity=0.85),
            secondary_y=False
        )
        
        fig3.add_trace(
            go.Scatter(name='Avg Speed (km/h)', 
                       x=severity_impact['Severity'],
                       y=severity_impact['Avg Speed'], 
                       mode='lines+markers',
                       marker=dict(size=14, color='#ef4444', symbol='diamond'),
                       line=dict(width=3, color='#ef4444')),
            secondary_y=True
        )
        
        fig3.update_layout(
            title='⚠️ Impact of Incident Severity on Traffic Conditions',
            template='plotly_white', 
            height=520,
            title_font_size=22,
            legend=dict(x=0.01, y=0.99, bgcolor='rgba(255,255,255,0.8)'),
            hovermode='x unified'
        )
        fig3.update_yaxes(title_text="Congestion Level", secondary_y=False, 
                          gridcolor='#e5e7eb')
        fig3.update_yaxes(title_text="Avg Speed (km/h)", secondary_y=True,
                          gridcolor='#fca5a5')
        fig3.update_xaxes(title_text="Incident Severity")
        fig3.show()
```

```python
# ========================================
# SECTION 4: DEVELOPED vs EMERGING MARKETS
# ========================================
print("\n🌍 Developed vs Emerging Markets Comparison")

# Classify cities directly using the known city_name column
developed = ['New York', 'London', 'Tokyo', 'Berlin', 'Sydney', 'Paris', 'Singapore']
emerging = ['Mumbai', 'São Paulo', 'Lagos', 'Jakarta', 'Cairo', 'Delhi', 'Bangalore']

cities['market'] = cities['city_name'].apply(
    lambda x: 'Developed' if x in developed else ('Emerging' if x in emerging else 'Other'))

# Merge and filter (drop_duplicates on just city_id to avoid the duplicate column issue)
city_map = cities[['city_id', 'city_name', 'market']].drop_duplicates(subset='city_id')

mkt = pd.merge(traffic, city_map, on='city_id', how='left')
mkt = mkt[mkt['market'].isin(['Developed', 'Emerging'])]

print(f"   Cities: {mkt['city_name'].nunique()} | Developed: {mkt[mkt['market']=='Developed']['city_name'].nunique()} | Emerging: {mkt[mkt['market']=='Emerging']['city_name'].nunique()}")

# Group and visualize
comp = mkt.groupby('market').agg(
    congestion=('congestion_level', 'mean'),
    speed=('avg_speed_kmh', 'mean')
).round(2).reset_index()

fig = make_subplots(1, 2, subplot_titles=('🏙️ Avg Congestion', '🚗 Avg Speed (km/h)'))
colors = {'Developed': '#3b82f6', 'Emerging': '#f59e0b'}

fig.add_trace(go.Bar(x=comp['market'], y=comp['congestion'], text=comp['congestion'],
                     textposition='outside', marker_color=[colors[m] for m in comp['market']]), row=1, col=1)
fig.add_trace(go.Bar(x=comp['market'], y=comp['speed'], text=comp['speed'],
                     textposition='outside', marker_color=[colors[m] for m in comp['market']]), row=1, col=2)

fig.update_layout(title='🌍 Developed vs Emerging Markets', template='plotly_white',
                  height=450, showlegend=False)
fig.update_yaxes(title_text="Congestion Level", row=1, col=1)
fig.update_yaxes(title_text="Speed (km/h)", row=1, col=2)
fig.show()

# Summary table
display(comp.style.set_properties(**{'text-align': 'center', 'font-size': '14px'})
        .background_gradient(cmap='Blues').format(precision=2))

print("\n✅ Market comparison complete!")
```
