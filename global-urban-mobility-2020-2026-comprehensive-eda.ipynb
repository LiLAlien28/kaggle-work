```python
# ========================================
# 2. DATA LOADING & INITIAL OVERVIEW
# ========================================

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print("🚀 Loading all dataset files...\n")

# Load all main files
cities = pd.read_parquet('/kaggle/input/datasets/aamir28/global-urban-mobility-and-smart-cities-dataset/cities_metadata.parquet')
traffic = pd.read_parquet('/kaggle/input/datasets/aamir28/global-urban-mobility-and-smart-cities-dataset/daily_traffic_metrics.parquet')
modes = pd.read_parquet('/kaggle/input/datasets/aamir28/global-urban-mobility-and-smart-cities-dataset/mobility_modes.parquet')
env = pd.read_parquet('/kaggle/input/datasets/aamir28/global-urban-mobility-and-smart-cities-dataset/environmental_factors.parquet')
incidents = pd.read_parquet('/kaggle/input/datasets/aamir28/global-urban-mobility-and-smart-cities-dataset/incidents_events.parquet')
agg = pd.read_parquet('/kaggle/input/datasets/aamir28/global-urban-mobility-and-smart-cities-dataset/aggregated_daily_indicators.parquet')

# Basic Information
print(f"✅ Successfully loaded {len(cities)} cities dataset")
print(f"   • Traffic records      : {len(traffic):,}")
print(f"   • Mobility modes rows  : {len(modes):,}")
print(f"   • Environmental records: {len(env):,}")
print(f"   • Incidents            : {len(incidents):,}")
print(f"   • Date Range           : {traffic['date'].min()} to {traffic['date'].max()}\n")

# City List
print("🌍 Cities Included:")
display(cities[['city_name', 'country', 'region', 'population_2025']].style.hide(axis="index"))

# Summary Table
summary = pd.DataFrame({
    'File': ['cities_metadata', 'daily_traffic_metrics', 'mobility_modes', 
             'environmental_factors', 'incidents_events', 'aggregated_daily_indicators'],
    'Rows': [len(cities), len(traffic), len(modes), len(env), len(incidents), len(agg)],
    'Columns': [cities.shape[1], traffic.shape[1], modes.shape[1], 
                env.shape[1], incidents.shape[1], agg.shape[1]],
    'Description': [
        'City metadata and coordinates',
        'Hourly traffic & congestion data',
        'Ride-hailing, transit & EV data',
        'Weather & air quality',
        'Incidents & disruptions',
        'Daily aggregated metrics'
    ]
})

print("\n📋 Dataset Files Summary:")
display(summary.style.set_properties(**{'text-align': 'left'}))
```

```python
# ========================================
# 3. EXPLORATORY DATA ANALYSIS - TRAFFIC PATTERNS
# ========================================

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Set style
plt.style.use('default')
sns.set_palette("husl")

print(" Creating Visualizations...\n")

# 1. Average Congestion by Hour of Day (Global)
hourly = traffic.groupby('hour_of_day')['congestion_level'].mean().reset_index()

fig1 = px.line(hourly, x='hour_of_day', y='congestion_level', 
               title=' Average Congestion Level by Hour of Day (All Cities)',
               labels={'hour_of_day': 'Hour of Day', 'congestion_level': 'Avg Congestion Level'},
               markers=True, template='plotly_white')

fig1.update_layout(height=500, title_font_size=20)
fig1.show()

# 2. Congestion Comparison by City
city_congestion = traffic.groupby('city_id')['congestion_level'].mean().reset_index()
city_congestion = city_congestion.merge(cities[['city_id', 'city_name', 'region']], on='city_id')

fig2 = px.bar(city_congestion.sort_values('congestion_level', ascending=False), 
              x='city_name', y='congestion_level', color='region',
              title=' Average Congestion Level by City',
              labels={'congestion_level': 'Average Congestion Level', 'city_name': 'City'},
              template='plotly_white')

fig2.update_layout(height=550, xaxis_tickangle=-45)
fig2.show()

# 3. Speed vs Congestion (Scatter)
fig3 = px.scatter(traffic.sample(15000, random_state=42), 
                  x='congestion_level', y='avg_speed_kmh',
                  color='city_id', opacity=0.6,
                  title='Speed vs Congestion Level (Sampled)',
                  labels={'congestion_level': 'Congestion Level', 
                          'avg_speed_kmh': 'Average Speed (km/h)'},
                  template='plotly_white')

fig3.show()

```
