## 3. Feature Engineering & Region Tags <a id='3'></a>


```python
REGION = {
    'Australia':'Asia-Pacific','New Zealand':'Asia-Pacific',
    'Japan':'Asia-Pacific','S. Korea':'Asia-Pacific',
    'Canada':'Americas','US':'Americas','Colombia':'Americas',
    'S. Africa':'Africa', 'Israel':'Middle East',
    'Germany':'Europe','France':'Europe','UK':'Europe','Italy':'Europe',
    'Spain':'Europe','Portugal':'Europe','Netherlands':'Europe',
    'Belgium':'Europe','Sweden':'Europe','Norway':'Europe',
    'Denmark':'Europe','Finland':'Europe','Ireland':'Europe',
    'Switzerland':'Europe','Luxembourg':'Europe',
    'Croatia':'Europe','Slovenia':'Europe',
}
REGION_PAL = {
    'Asia-Pacific':'#38bdf8','Americas':'#f472b6',
    'Europe':'#818cf8','Africa':'#34d399','Middle East':'#fb923c',
}

df['region']       = df['country'].map(REGION)
df['region_color'] = df['region'].map(REGION_PAL)

# HPI/PDI divergence: how much faster houses rose vs income
df['hpi_pdi_ratio'] = df['real_house_price_index'] / df['real_personal_disposable_income']

# Era flags
df['era'] = pd.cut(df['year'],
                    bins=[1974,1984,1994,2004,2014,2025],
                    labels=['1975-84','1985-94','1995-04','2005-14','2015-25'])

# Per-country: earliest 5yr vs latest 3yr affordability
countries = sorted(df['country'].unique())
aff_change = []
for c in countries:
    sub = df[df['country']==c].sort_values('year')
    early = sub.head(5)['affordability_proxy'].mean()
    late  = sub.tail(3)['affordability_proxy'].mean()
    aff_change.append({'country':c, 'early_ap':early,
                        'late_ap':late, 'delta_ap':late-early,
                        'region': REGION[c]})
aff_df = pd.DataFrame(aff_change).sort_values('delta_ap', ascending=False)
print('Affordability change summary:')
print(aff_df.to_string(index=False))
```

## 4. Affordability Timeline — All 26 Countries <a id='4'></a>


```python
fig, ax = plt.subplots(figsize=(18, 9))
fig.patch.set_facecolor(BG)
ax.set_facecolor(PANEL)

for country in countries:
    sub = df[df['country']==country].sort_values('year')
    color = REGION_PAL[REGION[country]]
    ax.plot(sub['year'], sub['affordability_proxy'],
            color=color, linewidth=1.1, alpha=0.55, zorder=3)
    # Label the endpoint
    last = sub.iloc[-1]
    ax.text(last['year']+0.3, last['affordability_proxy'],
            country.replace('S. ','S.'), fontsize=6.5,
            color=color, va='center', alpha=0.85)

# Global average overlay
yr_avg = df.groupby('year')['affordability_proxy'].mean()
ax.plot(yr_avg.index, yr_avg.values, color='white', linewidth=2.8,
        zorder=6, label='Global Average')

# Reference line
ax.axhline(100, color='#f59e0b', lw=1.5, ls='--', alpha=0.7,
           label='2024 Baseline (100)')

# Era shading
for yr, label, alpha in [(2007,'GFC 2007-09',0.06),(2020,'COVID Boom',0.06)]:
    ax.axvspan(yr, yr+2, color='#f87171', alpha=alpha)
    ax.text(yr+1, ax.get_ylim()[0] if ax.get_ylim()[0]>0 else 20,
            label, color='#f87171', fontsize=8, ha='center', alpha=0.8)

# Legend patches
patches = [mpatches.Patch(color=c, label=r) for r,c in REGION_PAL.items()]
patches += [Line2D([0],[0],color='white',lw=2.5,label='Global Avg'),
            Line2D([0],[0],color='#f59e0b',lw=1.5,ls='--',label='Baseline 100')]
ax.legend(handles=patches, fontsize=9, loc='upper right',
          ncol=2, framealpha=0.85)

ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Affordability Proxy  (>100 = less affordable than 2024)', fontsize=11)
ax.set_title('Housing Affordability Proxy 1975–2025 · All 26 Countries',
             fontsize=15, fontweight='bold')
ax.set_xlim(1975, 2027)
ax.grid(True, zorder=0); ax.set_axisbelow(True)

plt.tight_layout()
plt.savefig('02_affordability_timeline.png', dpi=150, bbox_inches='tight', facecolor=BG)
plt.show()
```

## 5. Who Got Worse? Affordability Δ 1975→2025 <a id='5'></a>


```python
fig, axes = plt.subplots(1, 2, figsize=(18, 8))
fig.patch.set_facecolor(BG)
fig.suptitle('Affordability Change: Early 1975–79 vs Late 2023–25',
             fontsize=16, fontweight='bold', color='white', y=1.01)

aff_sorted = aff_df.sort_values('delta_ap')

# 1. Horizontal bar — delta
ax = axes[0]
colors = ['#f87171' if d > 0 else '#34d399' for d in aff_sorted['delta_ap']]
bars = ax.barh(aff_sorted['country'], aff_sorted['delta_ap'],
               color=colors, height=0.7, zorder=3, edgecolor='none', alpha=0.88)
ax.axvline(0, color='white', lw=1, alpha=0.5)
ax.set_xlabel('Δ Affordability Proxy (positive = WORSE)')
ax.set_title('Affordability Deterioration (Δ)', fontweight='bold')
ax.grid(axis='x', zorder=0); ax.set_axisbelow(True)
for bar, val in zip(bars, aff_sorted['delta_ap']):
    x = bar.get_width()
    ax.text(x + (1.5 if x >= 0 else -1.5), bar.get_y()+bar.get_height()/2,
            f'{val:+.1f}', va='center', fontsize=8.5,
            color='#f87171' if val>0 else '#34d399')

# 2. Dumbbell: early vs late
ax = axes[1]
ax.set_facecolor(PANEL)
y_pos = range(len(aff_sorted))
for i, (_, row) in enumerate(aff_sorted.iterrows()):
    ax.plot([row['early_ap'], row['late_ap']], [i, i],
            color='#334155', linewidth=1.5, zorder=2)
    ax.scatter(row['early_ap'], i, color='#34d399', s=55, zorder=4)
    ax.scatter(row['late_ap'],  i, color='#f87171', s=55, zorder=4)

ax.set_yticks(list(y_pos))
ax.set_yticklabels(aff_sorted['country'], fontsize=9)
ax.set_xlabel('Affordability Proxy')
ax.set_title('Early (●) vs Late (●) Affordability', fontweight='bold')
ax.legend(handles=[
    Line2D([0],[0],marker='o',color='w',markerfacecolor='#34d399',ms=9,label='1975–79'),
    Line2D([0],[0],marker='o',color='w',markerfacecolor='#f87171',ms=9,label='2023–25'),
], fontsize=10, loc='lower right')
ax.grid(axis='x', zorder=0); ax.set_axisbelow(True)

plt.tight_layout()
plt.savefig('03_affordability_delta.png', dpi=150, bbox_inches='tight', facecolor=BG)
plt.show()
```

## 6. Real House Price Index: 50-Year Journey <a id='6'></a>


```python
# Spotlight on 8 countries with most dramatic trajectories
spotlight = ['Australia','Ireland','Luxembourg','New Zealand',
             'UK','US','Germany','Japan']
spot_colors = ['#38bdf8','#f472b6','#fb923c','#34d399',
               '#818cf8','#f59e0b','#a78bfa','#94a3b8']

fig, axes = plt.subplots(2, 4, figsize=(20, 10))
fig.patch.set_facecolor(BG)
fig.suptitle('Real House Price Index — Spotlight Countries (1975=baseline)',
             fontsize=16, fontweight='bold', color='white', y=1.01)

for ax, country, color in zip(axes.flat, spotlight, spot_colors):
    sub = df[df['country']==country].sort_values('year')
    ax.fill_between(sub['year'], sub['real_house_price_index'],
                    alpha=0.18, color=color)
    ax.plot(sub['year'], sub['real_house_price_index'],
            color=color, linewidth=2.2, zorder=5)
    ax.axhline(100, color='#4a6f8a', lw=1, ls='--', alpha=0.7)

    # Annotate peak
    peak_yr = sub.loc[sub['real_house_price_index'].idxmax(), 'year']
    peak_val = sub['real_house_price_index'].max()
    ax.scatter(peak_yr, peak_val, color='white', s=60, zorder=6)
    ax.text(peak_yr, peak_val+3, f'Peak\n{peak_yr}: {peak_val:.0f}',
            ha='center', fontsize=7.5, color='white')

    start = sub.iloc[0]['real_house_price_index']
    end   = sub.iloc[-1]['real_house_price_index']
    pct   = (end - start)/start*100
    ax.set_title(f'{country}  ({pct:+.0f}%)', fontweight='bold', color=color)
    ax.set_xlabel('Year', fontsize=9)
    ax.set_ylabel('Real HPI', fontsize=9)
    ax.grid(True, zorder=0); ax.set_axisbelow(True)

plt.tight_layout()
plt.savefig('04_real_hpi_spotlight.png', dpi=150, bbox_inches='tight', facecolor=BG)
plt.show()
```

## 7. Real Income vs Real House Prices — Divergence <a id='7'></a>


```python
# Show HPI vs PDI dual-axis for 6 countries
focus6 = ['Australia','UK','Germany','Ireland','US','Japan']
f6_col  = ['#38bdf8','#818cf8','#a78bfa','#f472b6','#f59e0b','#94a3b8']

fig, axes = plt.subplots(2, 3, figsize=(19, 11))
fig.patch.set_facecolor(BG)
fig.suptitle('Real House Price Index vs Real Disposable Income — Divergence',
             fontsize=16, fontweight='bold', color='white', y=1.01)

for ax, country, color in zip(axes.flat, focus6, f6_col):
    sub = df[df['country']==country].sort_values('year')

    ax.fill_between(sub['year'], sub['real_house_price_index'],
                    sub['real_personal_disposable_income'],
                    where=sub['real_house_price_index'] > sub['real_personal_disposable_income'],
                    interpolate=True, alpha=0.25, color='#f87171',
                    label='Prices > Income')
    ax.fill_between(sub['year'], sub['real_house_price_index'],
                    sub['real_personal_disposable_income'],
                    where=sub['real_house_price_index'] <= sub['real_personal_disposable_income'],
                    interpolate=True, alpha=0.2, color='#34d399',
                    label='Income ≥ Prices')

    ax.plot(sub['year'], sub['real_house_price_index'],
            color='#f87171', linewidth=2.2, label='Real HPI', zorder=5)
    ax.plot(sub['year'], sub['real_personal_disposable_income'],
            color='#34d399', linewidth=2.2, linestyle='--', label='Real PDI', zorder=5)

    ax.set_title(country, fontweight='bold', color=color)
    ax.set_xlabel('Year', fontsize=9)
    ax.set_ylabel('Index', fontsize=9)
    ax.legend(fontsize=7.5, loc='upper left')
    ax.grid(True, zorder=0); ax.set_axisbelow(True)

plt.tight_layout()
plt.savefig('05_divergence.png', dpi=150, bbox_inches='tight', facecolor=BG)
plt.show()
```

```python
epochs = [
    ('Pre-GFC Boom',  2002, 2007, '#f59e0b'),
    ('GFC Crash',     2007, 2012, '#f87171'),
    ('Recovery',      2012, 2019, '#34d399'),
    ('COVID Surge',   2019, 2022, '#f472b6'),
    ('Correction',    2022, 2025, '#818cf8'),
]

fig, axes = plt.subplots(1, len(epochs), figsize=(22, 8), sharey=False)
fig.patch.set_facecolor(BG)
fig.suptitle('Real HPI Change Across Market Epochs — 26 Countries',
             fontsize=16, fontweight='bold', color='white', y=1.01)

for ax, (name, yr_start, yr_end, color) in zip(axes, epochs):
    changes = []
    for c in countries:
        sub = df[df['country']==c].sort_values('year')
        v_start = sub[sub['year']==yr_start]['real_house_price_index']
        v_end   = sub[sub['year']==yr_end  ]['real_house_price_index']
        if len(v_start) and len(v_end):
            pct = (v_end.values[0] - v_start.values[0]) / v_start.values[0] * 100
            changes.append({'country':c,'change':pct})

    ch_df = pd.DataFrame(changes).sort_values('change')
    bar_colors = ['#f87171' if v>0 else '#34d399' for v in ch_df['change']]
    # For crash epoch flip colour logic
    if name == 'GFC Crash':
        bar_colors = ['#34d399' if v>0 else '#f87171' for v in ch_df['change']]

    ax.barh(ch_df['country'], ch_df['change'],
            color=bar_colors, height=0.7, zorder=3, edgecolor='none', alpha=0.85)
    ax.axvline(0, color='white', lw=0.8, alpha=0.4)
    ax.set_title(f'{name}\n{yr_start}→{yr_end}', fontweight='bold', color=color)
    ax.set_xlabel('Δ Real HPI (%)', fontsize=9)
    ax.tick_params(labelsize=7)
    ax.grid(axis='x', zorder=0); ax.set_axisbelow(True)

plt.tight_layout()
plt.savefig('07_crisis_epochs.png', dpi=150, bbox_inches='tight', facecolor=BG)
plt.show()
```

## 13. Key Takeaways

### 🚨 Most Worsened (1975→2025)
- **Australia** — AP surged from ~39 to 100 (+60 pts); Real HPI up **+368%**
- **New Zealand** — from 47 to 100 (+53 pts); incomes left far behind
- **Canada** — from 51 to 100 (+49 pts); post-COVID spike still unresolved
- **Luxembourg** — from 57 to 103 (+45 pts); **+515%** real HPI growth
- **UK & Ireland** — each >40 pts deterioration since 1975

### ✅ Most Improved or Stable
- **Germany** — AP fell from 170 → 102; real wages outpaced house prices for decades
- **Japan** — from 180 → 99; post-bubble deflation made housing more accessible
- **Spain** — dramatic improvement from 185 → 103 (though volatile)
- **S. Korea & Colombia** — both started extremely unaffordable, converged to ~100
- **US** — relatively stable (97 → 100) compared to Anglo peers

### 📈 Post-COVID Surge (2019–2022)
- Global avg affordability proxy rose from 96.7 (2015) to 108.3 (2022) — sharpest 3-year spike
- Stimulus payments + ultra-low rates + supply shortages drove the surge
- Australia, NZ, Canada and Portugal were hardest hit in this era
- 2023–24 saw partial corrections as rates rose, but prices remain elevated

### 📊 2025 Acute Crisis Countries
- **Portugal 116.3** — highest affordability proxy in 2025; real HPI +84% since 1975
- **Spain 109.0** — post-recovery surge, tight rental & buy markets
- **Croatia 108.5** — EU accession driving demand beyond wage growth
- **Denmark & Netherlands** — both above 104; supply-constrained markets

### 📐 HPI vs Income Divergence
- Globally, real house prices have decoupled from real incomes since the mid-1990s
- The HPI/PDI ratio crossed 1.0 globally around 2003 and has mostly stayed above it
- GFC briefly corrected the divergence (2008–2012) but COVID re-accelerated it
- Germany & Japan are the notable exceptions where income kept pace with prices

### 🌍 Regional Patterns
- **Asia-Pacific** — most dramatic long-run worsening (AUS, NZ); Japan opposite
- **Americas** — Canada far worse than US; Colombia dramatically improved
- **Europe** — highly heterogeneous: Germany stable, Portugal/Spain surging
- No region escaped the post-COVID affordability shock entirely

> ⚠️ **Note on Interpretation:** The affordability proxy is indexed to 2024=100. Values above 100 mean housing is less affordable than the 2024 baseline (prices higher relative to incomes). Countries starting with very high values (S. Korea ~514, Spain ~185, Japan ~180) experienced genuine structural improvements — not data anomalies. All indices are in real (inflation-adjusted) terms.
